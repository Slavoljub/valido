````python
from sqlalchemy import create_engine, text

engine = create_engine("postgresql://postgres:postgres@localhost:5432/valido")

ddl = """
CREATE TABLE IF NOT EXISTS dobavljaci (
    id SERIAL PRIMARY KEY,
    naziv TEXT NOT NULL,
    pib TEXT UNIQUE
);

CREATE TABLE IF NOT EXISTS racuni (
    id SERIAL PRIMARY KEY,
    broj_racuna TEXT NOT NULL UNIQUE,
    datum_izdavanja DATE,
    dobavljac_id INT REFERENCES dobavljaci(id)
);

CREATE TABLE IF NOT EXISTS kpr_transakcije (
    id SERIAL PRIMARY KEY,
    redni_broj TEXT,
    datum_knjizenja DATE,
    racun_id INT REFERENCES racuni(id),
    osnovica NUMERIC,
    iznos_pdv NUMERIC,
    ukupan_iznos NUMERIC,
    pretporez_odbitni NUMERIC,
    pretporez_neodbitni NUMERIC,
    uvoz NUMERIC,
    naknada_poljoprivreda NUMERIC
);
"""

with engine.begin() as conn:
    for stmt in ddl.split(";"):
        s = stmt.strip()
        if s:
            conn.execute(text(s))

print("Tabele spremne.")
````

````python
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from openai import OpenAI
from sqlalchemy import text

# 1) env + client
load_dotenv(Path.cwd().parent / ".env", override=True)
os.environ.pop("SSL_CERT_FILE", None)  # zbog ranijeg SSL problema
os.environ.pop("SSL_CERT_DIR", None)

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def analyze_kpr_financial_risk(engine, top_n=50, model="gpt-4o"):
    """
    Iz baze čita KPR podatke, šalje GPT-u i vraća strukturisanu analizu rizika.
    """

    query = text("""
        SELECT
            k.redni_broj,
            k.datum_knjizenja,
            r.broj_racuna,
            d.naziv AS dobavljac,
            d.pib,
            k.osnovica,
            k.iznos_pdv,
            k.ukupan_iznos,
            k.pretporez_odbitni,
            k.pretporez_neodbitni,
            k.uvoz,
            k.naknada_poljoprivreda
        FROM kpr_transakcije k
        JOIN racuni r ON r.id = k.racun_id
        JOIN dobavljaci d ON d.id = r.dobavljac_id
        ORDER BY k.ukupan_iznos DESC
        LIMIT :top_n
    """)

    with engine.begin() as conn:
        rows = conn.execute(query, {"top_n": top_n}).mappings().all()

    sample_data = [dict(r) for r in rows]

    prompt = f"""
Ti si senior finansijski analitičar za PDV/KPR kontrolu u Srbiji.

Zadatak:
1) Identifikuj potencijalne anomalije i rizike u KPR stavkama.
2) Posebno obrati pažnju na:
   - neuobičajen odnos osnovica ↔ PDV
   - velike iznose i koncentraciju kod pojedinih dobavljača
   - potencijalno sporne stavke za internu proveru

Vrati STROGO JSON sa ovim ključevima:
{{
  "summary": "kratak rezime",
  "top_risks": [
    {{"redni_broj": "...", "dobavljac": "...", "reason": "...", "risk_level": "low|medium|high"}}
  ],
  "supplier_concentration_comment": "...",
  "recommended_checks": ["...", "...", "..."]
}}

Podaci:
{json.dumps(sample_data, default=str, ensure_ascii=False)}
"""

    resp = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "Odgovaraj precizno, stručno i u JSON formatu."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.1
    )

    content = resp.choices[0].message.content

    # pokušaj JSON parse
    try:
        return json.loads(content)
    except Exception:
        return {"raw_output": content}
````

````python
result = analyze_kpr_financial_risk(engine, top_n=50, model="gpt-4o")
result
````

````python
import json

def _extract_json(text: str):
    """Pokuša da izvuče JSON čak i kad model vrati ```json ... ```."""
    if not text:
        return None

    t = text.strip()

    # Ako je u fenced bloku
    if "```json" in t:
        t = t.split("```json", 1)[1]
        t = t.split("```", 1)[0].strip()
    elif t.startswith("```"):
        t = t.strip("`").strip()

    try:
        return json.loads(t)
    except Exception:
        return None


def format_kpr_report_polite(result: dict) -> str:
    """
    Pretvara JSON rezultat u lep, formalan izveštaj na srpskom.
    """
    if not isinstance(result, dict):
        return "Poštovani,\n\nAnaliza nije dostupna u očekivanom formatu.\n"

    # Ako je raw_output, probaj da ga parsiraš
    if "raw_output" in result:
        parsed = _extract_json(result.get("raw_output", ""))
        if parsed:
            result = parsed
        else:
            return (
                "Poštovani,\n\n"
                "U nastavku je vraćen neformatiran odgovor modela:\n\n"
                f"{result.get('raw_output', '')}"
            )

    summary = result.get("summary", "Nije dostupan rezime analize.")
    top_risks = result.get("top_risks", [])
    concentration = result.get("supplier_concentration_comment", "Nije dostupan komentar o koncentraciji dobavljača.")
    checks = result.get("recommended_checks", [])

    lines = []
    lines.append("Poštovani,")
    lines.append("")
    lines.append("na osnovu izvršene analize KPR podataka, možemo da uvidimo sledeće:")
    lines.append("")
    lines.append(f"{summary}")
    lines.append("")
    lines.append("Ključni identifikovani rizici:")
    if top_risks:
        for i, r in enumerate(top_risks, 1):
            redni = r.get("redni_broj", "N/A")
            dob = r.get("dobavljac", "N/A")
            raz = r.get("reason", "N/A")
            lvl = r.get("risk_level", "N/A")
            lines.append(f"{i}. Redni broj {redni} | Dobavljač: {dob} | Rizik: {lvl.upper()}")
            lines.append(f"   Obrazloženje: {raz}")
    else:
        lines.append("- Nisu identifikovani izraženi rizici u posmatranom uzorku.")
    lines.append("")
    lines.append("Komentar koncentracije dobavljača:")
    lines.append(f"{concentration}")
    lines.append("")
    lines.append("Preporučene kontrolne aktivnosti:")
    if checks:
        for c in checks:
            lines.append(f"- {c}")
    else:
        lines.append("- Dodatne preporuke trenutno nisu dostupne.")
    lines.append("")
    lines.append("Srdačan pozdrav,")
    lines.append("AI finansijski analitičar")
    return "\n".join(lines)
````

````python
result = analyze_kpr_financial_risk(engine, top_n=50, model="gpt-4o")
report_text = format_kpr_report_polite(result)
print(report_text)
````

````python
def run_kpr_ai_analysis(engine, top_n=50, model="gpt-4o"):
    """
    1) povlači KPR iz baze
    2) radi GPT analizu
    3) vraća i strukturisan rezultat + lep izveštaj
    """
    raw_result = analyze_kpr_financial_risk(engine, top_n=top_n, model=model)
    pretty_report = format_kpr_report_polite(raw_result)

    return {
        "ok": True,
        "top_n": top_n,
        "model": model,
        "analysis": raw_result,
        "report_text": pretty_report
    }

# test
final_result = run_kpr_ai_analysis(engine, top_n=50, model="gpt-4o")
print(final_result["report_text"])
````

````python
import pandas as pd
import numpy as np
from sqlalchemy import text

def build_kpr_signals(engine, limit=500):
    q = text("""
        SELECT
            k.id as kpr_id,
            k.redni_broj,
            k.datum_knjizenja,
            r.broj_racuna,
            d.naziv as dobavljac,
            d.pib,
            k.osnovica,
            k.iznos_pdv,
            k.ukupan_iznos,
            k.pretporez_odbitni,
            k.pretporez_neodbitni,
            k.uvoz
        FROM kpr_transakcije k
        JOIN racuni r ON r.id = k.racun_id
        JOIN dobavljaci d ON d.id = r.dobavljac_id
        ORDER BY k.datum_knjizenja DESC
        LIMIT :limit
    """)
    with engine.begin() as conn:
        df = pd.DataFrame(conn.execute(q, {"limit": limit}).mappings().all())

    if df.empty:
        return df

    df["osnovica"] = pd.to_numeric(df["osnovica"], errors="coerce").fillna(0.0)
    df["iznos_pdv"] = pd.to_numeric(df["iznos_pdv"], errors="coerce").fillna(0.0)
    df["ukupan_iznos"] = pd.to_numeric(df["ukupan_iznos"], errors="coerce").fillna(0.0)

    # Rule flags
    df["flag_missing_pib"] = df["pib"].astype(str).str.strip().eq("")
    df["pdv_ratio"] = np.where(df["osnovica"] > 0, df["iznos_pdv"] / df["osnovica"], np.nan)
    df["flag_pdv_out_of_range"] = (df["pdv_ratio"] < 0.15) | (df["pdv_ratio"] > 0.25)

    # Statistical flag (IQR za ukupan_iznos)
    q1, q3 = df["ukupan_iznos"].quantile(0.25), df["ukupan_iznos"].quantile(0.75)
    iqr = q3 - q1 if q3 > q1 else 1.0
    high_thr = q3 + 1.5 * iqr
    df["flag_amount_outlier"] = df["ukupan_iznos"] > high_thr

    # Score
    df["risk_score"] = (
        df["flag_missing_pib"].astype(int) * 25 +
        df["flag_pdv_out_of_range"].astype(int) * 45 +
        df["flag_amount_outlier"].astype(int) * 30
    )

    return df.sort_values("risk_score", ascending=False).reset_index(drop=True)

signals_df = build_kpr_signals(engine, limit=500)
signals_df.head(10)
````

````python
import json
from openai import OpenAI
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def analyze_with_llm_structured(signals_df, top_n=30, model="gpt-4o"):
    top = signals_df.head(top_n).copy()

    payload = top[[
        "kpr_id","redni_broj","broj_racuna","dobavljac","pib",
        "osnovica","iznos_pdv","ukupan_iznos","pdv_ratio",
        "flag_missing_pib","flag_pdv_out_of_range","flag_amount_outlier","risk_score"
    ]].to_dict(orient="records")

    prompt = f"""
Ti si finansijski analitičar za KPR.
Ulaz su već izračunati signali rizika.
Vrati SAMO validan JSON sa ključevima:
summary, top_risks, recommended_checks.

Format:
{{
  "summary": "string",
  "top_risks": [
    {{"kpr_id": 0, "redni_broj": "string", "risk_level": "low|medium|high", "reason": "string"}}
  ],
  "recommended_checks": ["string", "string"]
}}

Podaci:
{json.dumps(payload, ensure_ascii=False, default=str)}
"""

    resp = client.chat.completions.create(
        model=model,
        temperature=0.1,
        messages=[
            {"role":"system","content":"Odgovaraj strogo kao validan JSON bez markdown-a."},
            {"role":"user","content":prompt}
        ]
    )

    content = resp.choices[0].message.content.strip()
    return json.loads(content)

llm_result = analyze_with_llm_structured(signals_df, top_n=30, model="gpt-4o")
llm_result
````

````python
def render_business_report(result: dict) -> str:
    summary = result.get("summary", "")
    top_risks = result.get("top_risks", [])
    checks = result.get("recommended_checks", [])

    lines = []
    lines.append("Poštovani,")
    lines.append("")
    lines.append("na osnovu izvršene analize KPR podataka, uočeno je sledeće.")
    lines.append("")
    lines.append(summary if summary else "Sažetak trenutno nije dostupan.")
    lines.append("")
    lines.append("Najvažnije rizične stavke su:")
    if top_risks:
        for i, r in enumerate(top_risks, 1):
            lines.append(
                f"{i}. Redni broj {r.get('redni_broj','N/A')} "
                f"(kpr_id: {r.get('kpr_id','N/A')}), nivo rizika: {r.get('risk_level','N/A').upper()}. "
                f"Razlog: {r.get('reason','N/A')}"
            )
    else:
        lines.append("Nisu izdvojene rizične stavke u analiziranom uzorku.")
    lines.append("")
    lines.append("Preporučene kontrolne aktivnosti:")
    if checks:
        for c in checks:
            lines.append(f"- {c}")
    else:
        lines.append("- Dodatne preporuke trenutno nisu dostupne.")
    lines.append("")
    lines.append("Srdačan pozdrav,")
    lines.append("AI finansijski analitičar")
    return "\n".join(lines)

final_report = render_business_report(llm_result)
print(final_report)
````

````python
from __future__ import annotations

import re
from pathlib import Path
from typing import Any

import pandas as pd


def process_kpr_document(file_path: str | Path) -> dict[str, Any]:
    """Process a KPR document and return structured extraction result.

    Pipeline:
        document -> OCR/text ingestion -> parsing/extraction -> validation -> dict(JSON-ready)
    """

    def _clean_text(value: Any) -> str | None:
        if pd.isna(value):
            return None
        text = str(value).replace("\n", " ").replace("\t", " ")
        text = re.sub(r"\s+", " ", text).strip()
        if text in {"", "//", "/ /", "/"}:
            return None
        return text

    def _to_number(value: Any) -> float:
        if value is None or (isinstance(value, float) and pd.isna(value)):
            return 0.0
        text = str(value).strip().replace(" ", "")
        if text == "" or text.lower() == "nan":
            return 0.0
        if "," in text and "." in text:
            text = text.replace(".", "").replace(",", ".")
        else:
            text = text.replace(",", ".")
        try:
            return float(text)
        except ValueError:
            return 0.0

    def _split_redni_datum(value: str | None) -> tuple[str | None, str | None]:
        if not value:
            return None, None
        match = re.match(r"^\s*(\d+)\s+(\d{2}/\d{2}/\d{4})\s*$", value)
        if match:
            return match.group(1), match.group(2)
        fallback = re.match(r"^\s*(\d+)\s*$", value)
        if fallback:
            return fallback.group(1), None
        return None, None

    def _split_racun_datum(value: str | None) -> tuple[str | None, str | None]:
        if not value:
            return None, None
        match = re.match(r"^\s*(.+?)\s+(\d{2}/\d{2}/\d{4})\s*$", value)
        if match:
            return match.group(1).strip(), match.group(2)
        return value, None

    def _load_document(path: Path) -> tuple[pd.DataFrame | None, str, str]:
        suffix = path.suffix.lower()
        if suffix in {".xlsx", ".xls"}:
            return pd.read_excel(path, header=None), "excel", ""

        if suffix == ".csv":
            return pd.read_csv(path, header=None), "csv", ""

        if suffix == ".pdf":
            raw_text = ""
            try:
                import pdfplumber  # type: ignore
                with pdfplumber.open(path) as pdf:
                    raw_text = "\n".join((page.extract_text() or "") for page in pdf.pages)
            except Exception:
                raw_text = ""

            ocr_engine = "pdfplumber"
            if not raw_text.strip():
                try:
                    import pytesseract  # type: ignore
                    from PIL import Image  # type: ignore
                    from pdf2image import convert_from_path  # type: ignore

                    pages = convert_from_path(str(path))
                    raw_text = "\n".join(
                        pytesseract.image_to_string(Image.fromarray(page)) for page in pages
                    )
                    ocr_engine = "pytesseract"
                except Exception:
                    ocr_engine = "unavailable"

            table_rows = []
            for line in raw_text.splitlines():
                line = line.strip()
                if line:
                    table_rows.append(re.split(r"\s{2,}", line))
            frame = pd.DataFrame(table_rows) if table_rows else None
            return frame, ocr_engine, raw_text

        raise ValueError(f"Unsupported file type: {suffix}")

    def _extract_from_table(raw: pd.DataFrame) -> pd.DataFrame:
        if raw.empty:
            return pd.DataFrame()

        candidate_index = raw[raw[0].astype(str).str.match(r"^\s*0{2,}\d", na=False)].index
        if len(candidate_index) > 0:
            raw = raw.iloc[candidate_index[0]:].reset_index(drop=True)

        raw = raw.dropna(axis=1, how="all")

        expected_columns = [
            "redni_datum", "dokument", "racun_datum", "dobavljac", "pib",
            "ukupan_iznos", "osnovica", "oslobodjeno_pdv", "naknada_bez_pdv", "pdv",
            "pretporez_odbitni", "pretporez_neodbitni", "uvoz", "vrednost_bez_pdv",
            "iznos_pdv", "naknada_poljoprivreda",
        ]

        for idx in range(len(raw.columns), len(expected_columns)):
            raw[idx] = None
        raw = raw.iloc[:, :len(expected_columns)].copy()
        raw.columns = expected_columns

        for col in ["redni_datum", "dokument", "racun_datum", "dobavljac", "pib"]:
            raw[col] = raw[col].map(_clean_text)

        raw = raw[
            ~raw["dobavljac"].fillna("").str.contains("MEĐUZBIR|MEDJUZBIR|UKUPNO", case=False, regex=True)
        ]

        raw[["redni_broj", "datum_knjizenja"]] = raw["redni_datum"].apply(lambda x: pd.Series(_split_redni_datum(x)))
        raw[["broj_racuna", "datum_izdavanja"]] = raw["racun_datum"].apply(lambda x: pd.Series(_split_racun_datum(x)))

        raw["datum_knjizenja"] = pd.to_datetime(raw["datum_knjizenja"], dayfirst=True, errors="coerce")
        raw["datum_izdavanja"] = pd.to_datetime(raw["datum_izdavanja"], dayfirst=True, errors="coerce")

        numeric_cols = [
            "ukupan_iznos", "osnovica", "oslobodjeno_pdv", "naknada_bez_pdv", "pdv",
            "pretporez_odbitni", "pretporez_neodbitni", "uvoz", "vrednost_bez_pdv",
            "iznos_pdv", "naknada_poljoprivreda",
        ]
        for col in numeric_cols:
            raw[col] = raw[col].map(_to_number)

        out = raw[
            [
                "redni_broj", "datum_knjizenja", "broj_racuna", "datum_izdavanja",
                "dobavljac", "pib", "osnovica", "iznos_pdv", "ukupan_iznos",
                "pretporez_odbitni", "pretporez_neodbitni", "uvoz", "naknada_poljoprivreda",
            ]
        ].copy()

        out["pib"] = out["pib"].astype(str).str.replace(r"\D", "", regex=True)
        out = out[out["redni_broj"].notna()].reset_index(drop=True)
        return out

    def _validate_rows(df: pd.DataFrame) -> tuple[list[dict[str, Any]], dict[str, Any]]:
        entries: list[dict[str, Any]] = []
        errors: list[dict[str, Any]] = []

        for idx, row in df.iterrows():
            calc_total = round(float(row["osnovica"]) + float(row["iznos_pdv"]), 2)
            total = round(float(row["ukupan_iznos"]), 2)
            diff = round(abs(calc_total - total), 2)
            row_errors = []

            pib = (row.get("pib") or "").strip()
            if pib and len(pib) != 9:
                row_errors.append("PIB must have 9 digits")

            if diff > 0.5:
                row_errors.append("Total mismatch: osnovica + iznos_pdv != ukupan_iznos")

            entries.append(
                {
                    "row_index": idx + 1,
                    "document_date": row["datum_izdavanja"].strftime("%Y-%m-%d") if pd.notna(row["datum_izdavanja"]) else None,
                    "invoice_number": row["broj_racuna"],
                    "supplier_name": row["dobavljac"],
                    "supplier_pib": pib if pib else None,
                    "tax_base_amount": float(row["osnovica"]),
                    "vat_amount": float(row["iznos_pdv"]),
                    "total_amount": float(row["ukupan_iznos"]),
                    "validation": {"is_valid": len(row_errors) == 0, "errors": row_errors},
                }
            )

            if row_errors:
                errors.append({"row_index": idx + 1, "errors": row_errors})

        summary = {
            "rows_total": len(df),
            "rows_valid": len(df) - len(errors),
            "rows_invalid": len(errors),
            "errors": errors,
        }
        return entries, summary

    path = Path(file_path)
    if not path.exists():
        raise FileNotFoundError(f"File not found: {path}")

    raw_df, ocr_engine, raw_text = _load_document(path)
    if raw_df is None or raw_df.empty:
        return {
            "document": {"file_path": str(path), "file_type": path.suffix.lower(), "ocr_engine": ocr_engine},
            "entries": [],
            "validation_summary": {
                "rows_total": 0,
                "rows_valid": 0,
                "rows_invalid": 0,
                "errors": ["No extractable rows found"],
            },
            "raw_text": raw_text,
        }

    extracted = _extract_from_table(raw_df)
    entries, validation_summary = _validate_rows(extracted)

    result = {
        "document": {
            "file_path": str(path),
            "file_type": path.suffix.lower(),
            "ocr_engine": ocr_engine,
            "rows_extracted": len(entries),
        },
        "entries": entries,
        "totals": {
            "tax_base_sum": round(float(extracted["osnovica"].sum()), 2) if not extracted.empty else 0.0,
            "vat_sum": round(float(extracted["iznos_pdv"].sum()), 2) if not extracted.empty else 0.0,
            "total_sum": round(float(extracted["ukupan_iznos"].sum()), 2) if not extracted.empty else 0.0,
        },
        "validation_summary": validation_summary,
        "raw_text": raw_text,
    }

    return result
````
