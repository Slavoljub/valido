```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import json

# Connection parameters
DB_CONFIG = {
    "host": "localhost",
    "port": 6379,
    "password": "redis123",
    "decode_responses": True
}

def connect_redis():
    """Connect to Redis database"""
    try:
        import redis
        r = redis.Redis(**DB_CONFIG)
        r.ping()
        print("✅ Connected to Redis successfully!")
        return r
    except Exception as e:
        print(f"❌ Connection failed: {e}")
        return None

def list_keys(redis_client):
    """List all keys"""
    return redis_client.keys("*")

def get_info(redis_client):
    """Get Redis server information"""
    return redis_client.info()

# Test connection
redis_client = connect_redis()
if redis_client:
    print(f"\n🔑 Keys: {list_keys(redis_client)}")
    print(f"📊 Info: {get_info(redis_client)}")
```
