```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import json

# Connection parameters
DB_CONFIG = {
    "host": "localhost",
    "port": 27017,
    "database": "ai_valido_online",
    "username": "admin",
    "password": "admin123"
}

def connect_mongodb():
    """Connect to MongoDB database"""
    try:
        from pymongo import MongoClient
        client = MongoClient(f"mongodb://{DB_CONFIG['username']}:{DB_CONFIG['password']}@{DB_CONFIG['host']}:{DB_CONFIG['port']}/")
        db = client[DB_CONFIG['database']]
        print("✅ Connected to MongoDB successfully!")
        return client, db
    except Exception as e:
        print(f"❌ Connection failed: {e}")
        return None, None

def list_databases(client):
    """List all databases"""
    return client.list_database_names()

def list_collections(db):
    """List all collections in current database"""
    return db.list_collection_names()

# Test connection
client, db = connect_mongodb()
if client and db:
    print(f"\n📊 Databases: {list_databases(client)}")
    print(f"📋 Collections: {list_collections(db)}")
```
