## 📄 PDF Processing and Document Analysis

Let's explore advanced PDF processing capabilities including text extraction, image extraction, and document analysis.

```python
# Install PDF processing packages
!pip install PyPDF2 pdfplumber reportlab python-docx openpyxl pytesseract tesseract-ocr

# Import PDF processing libraries
import PyPDF2
import pdfplumber
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet
import pytesseract
from PIL import Image
import io

print("✅ PDF processing packages installed and imported")
```

```python
class PDFProcessor:
    """Advanced PDF processor with OCR and table detection"""
    
    def __init__(self):
        self.ocr_available = True
        try:
            pytesseract.get_tesseract_version()
        except:
            self.ocr_available = False
            print("⚠️ Tesseract OCR not available")
    
    def extract_text(self, pdf_path, ocr_fallback=True):
        """Extract text from PDF with OCR fallback"""
        text = ""
        try:
            # Try with pdfplumber first (better for structured PDFs)
            with pdfplumber.open(pdf_path) as pdf:
                for page in pdf.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"
        except Exception as e:
            print(f"⚠️ pdfplumber failed: {e}, trying PyPDF2")
            try:
                # Fallback to PyPDF2
                with open(pdf_path, 'rb') as file:
                    pdf_reader = PyPDF2.PdfReader(file)
                    for page_num in range(len(pdf_reader.pages)):
                        page = pdf_reader.pages[page_num]
                        text += page.extract_text() + "\n"
            except Exception as e2:
                print(f"⚠️ PyPDF2 failed: {e2}")
                if ocr_fallback and self.ocr_available:
                    text = self._extract_with_ocr(pdf_path)
        
        return text.strip()
    
    def _extract_with_ocr(self, pdf_path):
        """Extract text using OCR from PDF images"""
        text = ""
        try:
            with pdfplumber.open(pdf_path) as pdf:
                for page in pdf.pages:
                    # Convert page to image
                    img = page.to_image()
                    if img:
                        # Extract text using OCR
                        ocr_text = pytesseract.image_to_string(img.original)
                        text += ocr_text + "\n"
        except Exception as e:
            print(f"⚠️ OCR extraction failed: {e}")
        
        return text.strip()
    
    def extract_images(self, pdf_path, output_dir="extracted_images"):
        """Extract all images from PDF"""
        images = []
        os.makedirs(output_dir, exist_ok=True)
        
        try:
            with pdfplumber.open(pdf_path) as pdf:
                for page_num, page in enumerate(pdf.pages):
                    # Extract images from page
                    page_images = page.images
                    for img_num, img in enumerate(page_images):
                        img_data = img['stream'].get_data()
                        img_filename = f"{output_dir}/page_{page_num+1}_img_{img_num+1}.png"
                        
                        with open(img_filename, 'wb') as f:
                            f.write(img_data)
                        images.append(img_filename)
                        print(f"✅ Extracted image: {img_filename}")
        except Exception as e:
            print(f"⚠️ Image extraction failed: {e}")
        
        return images
    
    def create_sample_pdf(self, output_path="sample_document.pdf"):
        """Create a sample PDF for testing"""
        doc = SimpleDocTemplate(output_path, pagesize=letter)
        styles = getSampleStyleSheet()
        content = []
        
        # Add title
        content.append(Paragraph("Sample PDF Document for Computer Vision Testing", styles['Title']))
        content.append(Spacer(1, 12))
        
        # Add text content
        content.append(Paragraph("This is a sample PDF document created for testing the advanced PDF processing capabilities.", styles['Normal']))
        content.append(Spacer(1, 12))
        
        # Add table
        data = [['Feature', 'Status', 'Description'],
                ['Text Extraction', '✅', 'Extract text from PDFs'],
                ['Image Extraction', '✅', 'Extract images from PDFs'],
                ['OCR Processing', '✅', 'Optical Character Recognition'],
                ['Table Detection', '✅', 'Detect and extract tables']]
        
        table = Table(data)
        table.setStyle(TableStyle([
            ('BACKGROUND', (0, 0), (-1, 0), '#4CAF50'),
            ('TEXTCOLOR', (0, 0), (-1, 0), 'white'),
            ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
            ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
            ('FONTSIZE', (0, 0), (-1, 0), 14),
            ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
            ('BACKGROUND', (0, 1), (-1, -1), '#f9f9f9'),
            ('GRID', (0, 0), (-1, -1), 1, 'black')
        ]))
        
        content.append(table)
        doc.build(content)
        print(f"✅ Created sample PDF: {output_path}")
        return output_path

# Initialize PDF processor
pdf_processor = PDFProcessor()
print("✅ PDF Processor initialized")
```

## 🎨 Advanced Image Processing and Creative Examples

Let's explore advanced image processing techniques including various formats, creative effects, and batch processing.

```python
# Install additional image processing packages
!pip install imageio imageio-ffmpeg scikit-image
!pip install opencv-contrib-python

# Import additional libraries
import imageio
from skimage import io, color, filters, feature, transform, morphology
from skimage.segmentation import slic, watershed
from skimage.feature import canny, corner_harris, blob_doh
from skimage.measure import label, regionprops
from scipy import ndimage
import cv2

print("✅ Advanced image processing packages installed")
```

```python
class AdvancedImageProcessor:
    """Advanced image processor with creative effects and multi-format support"""
    
    def __init__(self):
        self.supported_formats = ['.jpg', '.jpeg', '.png', '.tiff', '.bmp', '.webp', '.heic']
    
    def load_image_multi_format(self, image_path):
        """Load image in various formats"""
        try:
            # Try OpenCV first
            img = cv2.imread(image_path)
            if img is not None:
                return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            # Try PIL
            img = Image.open(image_path)
            return np.array(img)
            
        except Exception as e:
            print(f"⚠️ Failed to load image: {e}")
            return None
    
    def apply_creative_effects(self, image):
        """Apply various creative effects to image"""
        effects = {}
        
        # Convert to PIL for effects
        pil_img = Image.fromarray(image)
        
        # 1. Vintage effect
        vintage = pil_img.convert('RGB')
        vintage = ImageEnhance.Contrast(vintage).enhance(1.3)
        vintage = ImageEnhance.Color(vintage).enhance(0.8)
        vintage = ImageEnhance.Brightness(vintage).enhance(0.9)
        effects['vintage'] = np.array(vintage)
        
        # 2. Sketch effect
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        inverted = 255 - gray
        blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
        sketch = cv2.divide(gray, 255 - blurred, scale=256)
        effects['sketch'] = cv2.cvtColor(sketch, cv2.COLOR_GRAY2RGB)
        
        # 3. Oil painting effect
        try:
            oil_painting = cv2.xphoto.oilPainting(image, 7, 1)
            effects['oil_painting'] = oil_painting
        except:
            print("⚠️ Oil painting effect not available")
        
        # 4. Watercolor effect
        try:
            watercolor = cv2.stylization(image, sigma_s=60, sigma_r=0.4)
            effects['watercolor'] = watercolor
        except:
            print("⚠️ Watercolor effect not available")
        
        # 5. Cartoon effect
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        gray = cv2.medianBlur(gray, 5)
        edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
        color = cv2.bilateralFilter(image, 9, 300, 300)
        cartoon = cv2.bitwise_and(color, color, mask=edges)
        effects['cartoon'] = cartoon
        
        return effects
    
    def batch_process_images(self, image_paths, operation='resize', **kwargs):
        """Process multiple images in batch"""
        results = []
        
        for i, path in enumerate(image_paths):
            try:
                img = self.load_image_multi_format(path)
                if img is not None:
                    if operation == 'resize':
                        processed = cv2.resize(img, (kwargs.get('width', 224), kwargs.get('height', 224)))
                    elif operation == 'grayscale':
                        processed = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
                    elif operation == 'blur':
                        processed = cv2.GaussianBlur(img, (kwargs.get('kernel', 15), kwargs.get('kernel', 15)), 0)
                    else:
                        processed = img
                    
                    results.append({
                        'path': path,
                        'processed': processed,
                        'success': True
                    })
                else:
                    results.append({
                        'path': path,
                        'processed': None,
                        'success': False,
                        'error': 'Failed to load image'
                    })
            except Exception as e:
                results.append({
                    'path': path,
                    'processed': None,
                    'success': False,
                    'error': str(e)
                })
        
        return results
    
    def extract_features_advanced(self, image):
        """Extract advanced features from image"""
        features = {}
        
        # Convert to grayscale for feature extraction
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        
        # 1. SIFT features
        sift = cv2.SIFT_create()
        keypoints, descriptors = sift.detectAndCompute(gray, None)
        features['sift'] = {
            'keypoints': len(keypoints),
            'descriptors': descriptors.shape if descriptors is not None else None
        }
        
        # 2. ORB features
        orb = cv2.ORB_create()
        keypoints, descriptors = orb.detectAndCompute(gray, None)
        features['orb'] = {
            'keypoints': len(keypoints),
            'descriptors': descriptors.shape if descriptors is not None else None
        }
        
        # 3. Corner detection
        corners = corner_harris(gray)
        features['corners'] = np.sum(corners > 0.01 * corners.max())
        
        # 4. Blob detection
        blobs = blob_doh(gray)
        features['blobs'] = len(blobs)
        
        # 5. Edge density
        edges = canny(gray)
        features['edge_density'] = np.sum(edges) / edges.size
        
        return features

# Initialize advanced image processor
advanced_processor = AdvancedImageProcessor()
print("✅ Advanced Image Processor initialized")
```

```python
class VideoProcessor:
    """Advanced video processor with effects and analysis"""
    
    def __init__(self):
        self.supported_formats = ['.mp4', '.avi', '.mov', '.mkv', '.wmv', '.flv']
    
    def extract_frames(self, video_path, output_dir="extracted_frames", frame_rate=1):
        """Extract frames from video at specified rate"""
        try:
            video = VideoFileClip(video_path)
            frames = []
            
            # Create output directory
            os.makedirs(output_dir, exist_ok=True)
            
            # Extract frames
            for i, frame in enumerate(video.iter_frames()):
                if i % int(video.fps / frame_rate) == 0:
                    frame_path = f"{output_dir}/frame_{i:04d}.jpg"
                    imageio.imwrite(frame_path, frame)
                    frames.append(frame_path)
            
            video.close()
            print(f"✅ Extracted {len(frames)} frames from video")
            return frames
            
        except Exception as e:
            print(f"⚠️ Frame extraction failed: {e}")
            return []
    
    def apply_video_effects(self, video_path, effects=['resize', 'fade']):
        """Apply effects to video"""
        try:
            video = VideoFileClip(video_path)
            processed_video = video
            
            for effect in effects:
                if effect == 'resize':
                    processed_video = processed_video.resize(width=640)
                elif effect == 'fade':
                    processed_video = processed_video.fadein(1).fadeout(1)
                elif effect == 'rotate':
                    processed_video = processed_video.rotate(90)
                elif effect == 'crop':
                    processed_video = processed_video.crop(x1=100, y1=100, x2=540, y2=380)
            
            # Save processed video
            output_path = video_path.replace('.mp4', '_processed.mp4')
            processed_video.write_videofile(output_path, codec='libx264')
            processed_video.close()
            video.close()
            
            print(f"✅ Applied effects: {effects}")
            return output_path
            
        except Exception as e:
            print(f"⚠️ Video effects failed: {e}")
            return None
    
    def detect_motion(self, video_path, threshold=30):
        """Detect motion in video"""
        try:
            video = VideoFileClip(video_path)
            motion_frames = []
            
            prev_frame = None
            
            for i, frame in enumerate(video.iter_frames()):
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                
                if prev_frame is not None:
                    # Calculate frame difference
                    frame_diff = cv2.absdiff(prev_frame, gray)
                    _, thresh = cv2.threshold(frame_diff, threshold, 255, cv2.THRESH_BINARY)
                    
                    # Calculate motion percentage
                    motion_percent = (np.sum(thresh) / thresh.size) * 100
                    
                    if motion_percent > 1:  # Motion detected
                        motion_frames.append({
                            'frame': i,
                            'motion_percent': motion_percent,
                            'timestamp': i / video.fps
                        })
                
                prev_frame = gray
            
            video.close()
            print(f"✅ Detected motion in {len(motion_frames)} frames")
            return motion_frames
            
        except Exception as e:
            print(f"⚠️ Motion detection failed: {e}")
            return []

# Initialize video processor
video_processor = VideoProcessor()
print("✅ Video Processor initialized")
```
