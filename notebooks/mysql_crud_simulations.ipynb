```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import json

# Connection parameters
DB_CONFIG = {
    "host": "localhost",
    "port": 3306,
    "database": "ai_valido_online",
    "user": "root",
    "password": "root"
}

def connect_mysql():
    """Connect to MySQL database"""
    try:
        import mysql.connector
        conn = mysql.connector.connect(**DB_CONFIG)
        print("✅ Connected to MySQL successfully!")
        return conn
    except Exception as e:
        print(f"❌ Connection failed: {e}")
        return None

def list_databases(conn):
    """List all databases"""
    cursor = conn.cursor()
    cursor.execute("SHOW DATABASES;")
    databases = cursor.fetchall()
    cursor.close()
    return [db[0] for db in databases]

def list_tables(conn):
    """List all tables in current database"""
    cursor = conn.cursor()
    cursor.execute("SHOW TABLES;")
    tables = cursor.fetchall()
    cursor.close()
    return [table[0] for table in tables]

def list_stored_procedures(conn):
    """List all stored procedures"""
    cursor = conn.cursor()
    cursor.execute("SHOW PROCEDURE STATUS WHERE Db = DATABASE();")
    procedures = cursor.fetchall()
    cursor.close()
    return [proc[1] for proc in procedures]

# Test connection
conn = connect_mysql()
if conn:
    print(f"\n📊 Databases: {list_databases(conn)}")
    print(f"📋 Tables: {list_tables(conn)}")
    print(f"🔧 Stored Procedures: {list_stored_procedures(conn)}")
```

```python
# Database Information Functions
def list_functions(conn):
    """List all functions"""
    cursor = conn.cursor()
    cursor.execute("SHOW FUNCTION STATUS WHERE Db = DATABASE();")
    functions = cursor.fetchall()
    cursor.close()
    return [func[1] for func in functions]

def list_views(conn):
    """List all views"""
    cursor = conn.cursor()
    cursor.execute("SHOW FULL TABLES WHERE Table_type = 'VIEW';")
    views = cursor.fetchall()
    cursor.close()
    return [view[0] for view in views]

if conn:
    print(f"🔧 Functions: {list_functions(conn)}")
    print(f"👁️ Views: {list_views(conn)}")
```
