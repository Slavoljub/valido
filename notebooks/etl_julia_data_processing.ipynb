## **🗄️ Multi-Database Integration**

Connect to and integrate data from all configured database sources.

```julia
# Multi-Database Connection Manager
struct DatabaseConnection
    db_type::String
    connection::Any
    config::Dict{String, Any}
end

function create_database_connections()
    connections = Dict{String, DatabaseConnection}()
    
    # PostgreSQL Connection
    try
        if get(ENV, "POSTGRES_ENABLED", "true") == "true"
            conn_string = "host=$(get(ENV, "POSTGRES_HOST", "localhost")) " *
                         "port=$(get(ENV, "POSTGRES_PORT", "5432")) " *
                         "dbname=$(get(ENV, "POSTGRES_NAME", "ai_valido_online")) " *
                         "user=$(get(ENV, "POSTGRES_USER", "postgres")) " *
                         "password=$(get(ENV, "POSTGRES_PASSWORD", "postgres"))"
            
            conn = LibPQ.Connection(conn_string)
            connections["postgresql"] = DatabaseConnection(
                "postgresql", conn, Dict("connection_string" => conn_string)
            )
            println("✅ PostgreSQL connection established")
        end
    catch e
        println("⚠️  PostgreSQL connection failed: ", e)
    end
    
    # MySQL Connection
    try
        if get(ENV, "MYSQL_ENABLED", "true") == "true"
            conn = MySQL.connect(
                get(ENV, "MYSQL_HOST", "localhost"),
                get(ENV, "MYSQL_USER", "root"),
                get(ENV, "MYSQL_PASSWORD", "root"),
                db=get(ENV, "MYSQL_NAME", "ai_valido_online")
            )
            connections["mysql"] = DatabaseConnection(
                "mysql", conn, Dict("host" => get(ENV, "MYSQL_HOST", "localhost"))
            )
            println("✅ MySQL connection established")
        end
    catch e
        println("⚠️  MySQL connection failed: ", e)
    end
    
    # MongoDB Connection
    try
        if get(ENV, "MONGODB_ENABLED", "true") == "true"
            client = Mongoc.Client(get(ENV, "MONGODB_URL", "mongodb://localhost:27017"))
            connections["mongodb"] = DatabaseConnection(
                "mongodb", client, Dict("url" => get(ENV, "MONGODB_URL", ""))
            )
            println("✅ MongoDB connection established")
        end
    catch e
        println("⚠️  MongoDB connection failed: ", e)
    end
    
    # Redis Connection
    try
        if get(ENV, "REDIS_ENABLED", "true") == "true"
            conn = Redis.Connection(
                host=get(ENV, "REDIS_HOST", "localhost"),
                port=parse(Int, get(ENV, "REDIS_PORT", "6379"))
            )
            connections["redis"] = DatabaseConnection(
                "redis", conn, Dict("host" => get(ENV, "REDIS_HOST", "localhost"))
            )
            println("✅ Redis connection established")
        end
    catch e
        println("⚠️  Redis connection failed: ", e)
    end
    
    # SQLite Connection (always available)
    try
        db_path = get(ENV, "SQLITE_PATH", "../data/sqlite/app.db")
        conn = SQLite.DB(db_path)
        connections["sqlite"] = DatabaseConnection(
            "sqlite", conn, Dict("path" => db_path)
        )
        println("✅ SQLite connection established")
    catch e
        println("⚠️  SQLite connection failed: ", e)
    end
    
    return connections
end

# Create all database connections
db_connections = create_database_connections()
println("\n📊 Database Connections Summary:")
for (db_type, conn) in db_connections
    println("  $db_type: ✅ Connected")
end
```

```julia
# Universal Data Extractor
function extract_data_from_sources(connections::Dict{String, DatabaseConnection}, query::String)
    results = Dict{String, DataFrame}()
    
    # PostgreSQL/MySQL Query
    for db_type in ["postgresql", "mysql"]
        if haskey(connections, db_type)
            try
                if db_type == "postgresql"
                    result = execute(connections[db_type].connection, query)
                    df = DataFrame(result)
                    results[db_type] = df
                    println("✅ Extracted $(nrow(df)) rows from PostgreSQL")
                elseif db_type == "mysql"
                    result = MySQL.query(connections[db_type].connection, query)
                    df = DataFrame(result)
                    results[db_type] = df
                    println("✅ Extracted $(nrow(df)) rows from MySQL")
                end
            catch e
                println("⚠️  $db_type query failed: ", e)
            end
        end
    end
    
    # MongoDB Query
    if haskey(connections, "mongodb")
        try
            collection = connections["mongodb"].connection["ai_valido_online"]["documents"]
            documents = collect(collection.find(Dict()))
            if !isempty(documents)
                df = DataFrame(documents)
                results["mongodb"] = df
                println("✅ Extracted $(nrow(df)) documents from MongoDB")
            end
        catch e
            println("⚠️  MongoDB query failed: ", e)
        end
    end
    
    # Redis Data (if available)
    if haskey(connections, "redis")
        try
            keys = Redis.keys(connections["redis"].connection, "*")
            redis_data = Dict()
            for key in keys
                value = Redis.get(connections["redis"].connection, key)
                redis_data[key] = value
            end
            if !isempty(redis_data)
                df = DataFrame(keys=collect(keys(redis_data)), values=collect(values(redis_data)))
                results["redis"] = df
                println("✅ Extracted $(nrow(df)) key-value pairs from Redis")
            end
        catch e
            println("⚠️  Redis query failed: ", e)
        end
    end
    
    return results
end

# Extract data from all sources
sample_query = "SELECT * FROM companies LIMIT 100"  # Adjust based on your schema
extracted_data = extract_data_from_sources(db_connections, sample_query)

println("\n📋 Data Extraction Summary:")
for (source, df) in extracted_data
    println("  $source: $(size(df, 1)) rows × $(size(df, 2)) columns")
end
```

## **📊 Advanced Data Visualization & Analytics**

Create comprehensive visualizations and perform advanced analytics on the integrated data.

```julia
# Advanced Data Visualization
function create_comprehensive_visualization(data_dict::Dict{String, DataFrame})
    plots = []
    
    # Create individual plots for each data source
    for (source, df) in data_dict
        if nrow(df) > 0
            try
                # Determine numeric columns for visualization
                numeric_cols = []
                for col in names(df)
                    if eltype(df[!, col]) <: Union{Number, Missing}
                        push!(numeric_cols, col)
                    end
                end
                
                if length(numeric_cols) >= 2
                    # Create scatter plot
                    x_col = numeric_cols[1]
                    y_col = numeric_cols[2]
                    
                    p = Plots.scatter(df[!, x_col], df[!, y_col],
                                    title="$source: $x_col vs $y_col",
                                    xlabel=string(x_col),
                                    ylabel=string(y_col),
                                    legend=:topright,
                                    markersize=4,
                                    alpha=0.7)
                    push!(plots, p)
                    
                elseif length(numeric_cols) >= 1
                    # Create histogram
                    col = numeric_cols[1]
                    p = Plots.histogram(df[!, col],
                                      title="$source: $col Distribution",
                                      xlabel=string(col),
                                      ylabel="Frequency",
                                      legend=false,
                                      bins=30)
                    push!(plots, p)
                end
                
            catch e
                println("⚠️  Visualization failed for $source: ", e)
            end
        end
    end
    
    return plots
end

# Create statistical analysis
function perform_statistical_analysis(data_dict::Dict{String, DataFrame})
    analysis_results = Dict{String, Dict{String, Any}}()
    
    for (source, df) in data_dict
        if nrow(df) > 0
            stats = Dict{String, Any}()
            
            # Basic statistics
            stats["row_count"] = nrow(df)
            stats["column_count"] = ncol(df)
            stats["column_names"] = names(df)
            
            # Column type analysis
            column_types = Dict()
            for col in names(df)
                column_types[col] = string(eltype(df[!, col]))
            end
            stats["column_types"] = column_types
            
            # Numeric column statistics
            numeric_stats = Dict()
            for col in names(df)
                if eltype(df[!, col]) <: Union{Number, Missing}
                    col_data = collect(skipmissing(df[!, col]))
                    if !isempty(col_data)
                        numeric_stats[col] = Dict(
                            "mean" => mean(col_data),
                            "median" => median(col_data),
                            "std" => std(col_data),
                            "min" => minimum(col_data),
                            "max" => maximum(col_data)
                        )
                    end
                end
            end
            stats["numeric_statistics"] = numeric_stats
            
            analysis_results[source] = stats
        end
    end
    
    return analysis_results
end

# Perform analysis if data is available
if !isempty(extracted_data)
    # Create visualizations
    println("📊 Creating visualizations...")
    plots = create_comprehensive_visualization(extracted_data)
    
    # Display plots
    if !isempty(plots)
        display(plots[1])  # Display first plot
        println("✅ Visualizations created successfully")
    end
    
    # Perform statistical analysis
    println("\n📈 Performing statistical analysis...")
    analysis_results = perform_statistical_analysis(extracted_data)
    
    # Display analysis summary
    println("\n📋 Statistical Analysis Summary:")
    for (source, stats) in analysis_results
        println("\n📊 $source Analysis:")
        println("  Rows: $(stats["row_count"])")
        println("  Columns: $(stats["column_count"])")
        if !isempty(stats["numeric_statistics"])
            println("  Numeric columns with stats: $(length(stats["numeric_statistics"]))")
        end
    end
end
```

## **📁 Multi-Format File Processing**

```julia
# Multi-Format File Processor
function process_file_by_format(file_path::String, format::String)
    try
        if !isfile(file_path)
            return "File not found: $file_path"
        end
        
        if format == "csv"
            df = CSV.read(file_path, DataFrame)
            return Dict("data" => df, "rows" => nrow(df), "cols" => ncol(df))
            
        elseif format == "json"
            json_data = JSON.parsefile(file_path)
            return Dict("data" => json_data, "type" => "json")
            
        elseif format == "xml"
            xml_data = XMLDict.parse_file(file_path)
            return Dict("data" => xml_data, "type" => "xml")
            
        elseif format == "excel" || format == "xlsx"
            df = XLSX.readtable(file_path, "Sheet1")
            return Dict("data" => DataFrame(df...), "rows" => size(df[2], 1), "cols" => size(df[2], 2))
            
        elseif format == "pdf"
            # This would require PDFIO.jl package
            return "PDF processing would be implemented here"
            
        elseif format == "parquet"
            df = Parquet.read_parquet(file_path)
            return Dict("data" => df, "rows" => nrow(df), "cols" => ncol(df))
            
        elseif format == "arrow"
            df = Arrow.Table(file_path)
            return Dict("data" => DataFrame(df), "rows" => size(df, 1), "cols" => size(df, 2))
            
        elseif format == "hdf5"
            data = HDF5.h5open(file_path, "r") do file
                read(file)
            end
            return Dict("data" => data, "type" => "hdf5")
            
        elseif format == "jld2"
            data = JLD2.load(file_path)
            return Dict("data" => data, "type" => "jld2")
            
        else
            return "Unsupported format: $format"
        end
        
    catch e
        return "Error processing $format file: $(e)"
    end
end

# Process sample files from configured paths
function process_sample_files()
    sample_files = [
        (get(ENV, "DOBAVLJACI_PATH", "../data/mom21/specifikacija_dobavljaca/specifikacija_dobavljaca_ai.csv"), "csv"),
        (get(ENV, "PLATE_PATH", "../data/mom21/plate/plate_2024.csv"), "csv"),
        (get(ENV, "SALARY_CSV_PATH", "../data/mom21/plate/mom21_plate_2024_bruto2_sorted.csv"), "csv")
    ]
    
    results = Dict{String, Any}()
    
    for (file_path, format) in sample_files
        println("📁 Processing $format file: $(basename(file_path))")
        result = process_file_by_format(file_path, format)
        results[basename(file_path)] = result
        
        if result isa Dict && haskey(result, "rows")
            println("  ✅ $(result["rows"]) rows × $(result["cols"]) columns")
        elseif result isa String
            println("  ⚠️  $result")
        end
    end
    
    return results
end

# Process sample files
file_processing_results = process_sample_files()

println("\n📁 File Processing Summary:")
for (filename, result) in file_processing_results
    if result isa Dict && haskey(result, "rows")
        println("  ✅ $filename: $(result["rows"]) rows")
    else
        println("  ⚠️  $filename: $result")
    end
end
```

## **⚡ Parallel & Distributed Processing**

Implement parallel processing for large-scale data operations.

```julia
# Parallel ETL Processing
function parallel_data_processing(data_sources::Vector{String}, processing_function::Function)
    # Use Julia's parallel processing capabilities
    using Distributed
    using Threads
    
    # Add worker processes if needed
    if nprocs() < length(data_sources)
        addprocs(length(data_sources) - nprocs())
    end
    
    println("🚀 Starting parallel processing with $(nprocs()) processes")
    println("📊 Processing $(length(data_sources)) data sources")
    
    # Parallel processing using @distributed
    results = @sync begin
        @distributed (vcat) for source in data_sources
            try
                result = processing_function(source)
                println("✅ Processed: $source")
                result
            catch e
                println("⚠️  Failed to process $source: $e")
                Dict("source" => source, "error" => string(e))
            end
        end
    end
    
    return results
end

# Define processing function for each data source
function process_single_source(source::String)
    # Simulate processing time
    sleep(rand() * 2)  # Random processing time up to 2 seconds
    
    # Mock data processing result
    return Dict(
        "source" => source,
        "processed_records" => rand(100:10000),
        "processing_time" => rand(),
        "status" => "completed",
        "timestamp" => now()
    )
end

# Define data sources to process in parallel
data_sources = [
    "postgresql://companies",
    "mysql://invoices",
    "mongodb://documents",
    "redis://cache",
    "elasticsearch://search",
    "csv://../data/mom21/plate/plate_2024.csv",
    "json://../data/embeddings/company_embeddings.json",
    "api://openai",
    "api://anthropic",
    "api://google"
]

# Run parallel processing
println("\n🔄 Starting Parallel ETL Processing...")
parallel_results = parallel_data_processing(data_sources, process_single_source)

# Analyze parallel processing results
successful_processes = filter(x -> get(x, "status", "") == "completed", parallel_results)
failed_processes = filter(x -> haskey(x, "error"), parallel_results)

println("\n📊 Parallel Processing Results:")
println("  ✅ Successful: $(length(successful_processes))")
println("  ❌ Failed: $(length(failed_processes))")

if !isempty(successful_processes)
    total_records = sum(get(x, "processed_records", 0) for x in successful_processes)
    avg_time = mean(get(x, "processing_time", 0.0) for x in successful_processes)
    
    println("  📈 Total Records Processed: $total_records")
    println("  ⏱️  Average Processing Time: $(round(avg_time, digits=3))s")
end
```

## **🧠 Machine Learning & Predictive Analytics**

Implement ML models for predictive analytics and data insights.

```julia
# Machine Learning Pipeline
function create_ml_pipeline(data::DataFrame)
    # Prepare data for ML
    println("🧠 Preparing data for machine learning...")
    
    # Identify numeric columns
    numeric_cols = []
    for col in names(data)
        if eltype(data[!, col]) <: Union{Number, Missing}
            push!(numeric_cols, col)
        end
    end
    
    if length(numeric_cols) < 2
        return "Insufficient numeric data for ML analysis (need at least 2 numeric columns)"
    end
    
    # Prepare features and target
    X = Matrix(data[!, numeric_cols[1:end-1]])  # Features
    y = Vector(data[!, numeric_cols[end]])       # Target
    
    # Remove missing values
    valid_indices = .!ismissing.(y)
    X = X[valid_indices, :]
    y = y[valid_indices]
    
    if size(X, 1) < 10
        return "Insufficient data for training (need at least 10 samples)"
    end
    
    # Split data
    n_samples = size(X, 1)
    n_train = Int(floor(n_samples * 0.7))
    
    X_train = X[1:n_train, :]
    y_train = y[1:n_train]
    X_test = X[n_train+1:end, :]
    y_test = y[n_train+1:end]
    
    println("📊 Data split: $(size(X_train, 1)) train, $(size(X_test, 1)) test samples")
    
    # Train multiple models
    models = Dict{String, Any}()
    predictions = Dict{String, Vector}()
    
    # Linear Regression
    try
        using MLJ
        
        # Linear Regression
        lin_reg = MLJ.LinearRegressor()
        MLJ.fit!(lin_reg, X_train, y_train)
        y_pred_lr = MLJ.predict(lin_reg, X_test)
        models["linear_regression"] = lin_reg
        predictions["linear_regression"] = y_pred_lr
        
        # Random Forest (if available)
        try
            rf_reg = MLJ.RandomForestRegressor()
            MLJ.fit!(rf_reg, X_train, y_train)
            y_pred_rf = MLJ.predict(rf_reg, X_test)
            models["random_forest"] = rf_reg
            predictions["random_forest"] = y_pred_rf
        catch
            println("⚠️  Random Forest not available")
        end
        
        println("✅ ML models trained successfully")
        
    catch e
        println("⚠️  ML training failed: ", e)
        return "ML training failed: $(e)"
    end
    
    # Evaluate models
    evaluation_results = Dict{String, Dict}()
    
    for (model_name, y_pred) in predictions
        try
            mse = mean((y_pred - y_test).^2)
            rmse = sqrt(mse)
            mae = mean(abs.(y_pred - y_test))
            r2 = 1 - sum((y_test - y_pred).^2) / sum((y_test - mean(y_test)).^2)
            
            evaluation_results[model_name] = Dict(
                "mse" => mse,
                "rmse" => rmse,
                "mae" => mae,
                "r2" => r2
            )
            
            println("📊 $model_name - R²: $(round(r2, digits=3)), RMSE: $(round(rmse, digits=3))")
            
        catch e
            println("⚠️  Evaluation failed for $model_name: ", e)
        end
    end
    
    return Dict(
        "models" => models,
        "predictions" => predictions,
        "evaluation" => evaluation_results,
        "feature_names" => numeric_cols[1:end-1],
        "target_name" => numeric_cols[end]
    )
end

# Apply ML to available data
ml_results = Dict{String, Any}()

for (source, df) in extracted_data
    if nrow(df) >= 20  # Need minimum data for ML
        println("\n🧠 Training ML models on $source data...")
        result = create_ml_pipeline(df)
        
        if result isa Dict
            ml_results[source] = result
            println("✅ ML pipeline completed for $source")
        else
            println("⚠️  $result")
        end
    else
        println("⚠️  Insufficient data for ML on $source ($(nrow(df)) rows)")
    end
end

println("\n🧠 Machine Learning Summary:")
for (source, result) in ml_results
    if haskey(result, "evaluation")
        model_count = length(result["evaluation"])
        println("  ✅ $source: $model_count models trained")
    end
end
```

## **📤 Export Results & Generate Reports**

Export processed data and generate comprehensive reports in multiple formats.

```julia
# Multi-Format Export Function
function export_data(data::Union{DataFrame, Dict}, filename::String, formats::Vector{String})
    base_name = splitext(filename)[1]
    results = Dict{String, String}()
    
    for format in formats
        try
            output_file = "$(base_name)_$(format).$(format)"
            
            if format == "csv"
                if data isa DataFrame
                    CSV.write(output_file, data)
                end
                
            elseif format == "json"
                if data isa DataFrame
                    json_data = Dict(zip(names(data), eachcol(data)))
                    JSON.print(IOBuffer(), json_data)
                elseif data isa Dict
                    open(output_file, "w") do f
                        JSON.print(f, data)
                    end
                end
                
            elseif format == "xml"
                # Simple XML export
                open(output_file, "w") do f
                    println(f, "<data>")
                    if data isa DataFrame
                        for row in eachrow(data)
                            println(f, "  <row>")
                            for (col, val) in pairs(row)
                                println(f, "    <$col>$val</$col>")
                            end
                            println(f, "  </row>")
                        end
                    end
                    println(f, "</data>")
                end
                
            elseif format == "excel"
                if data isa DataFrame
                    XLSX.writetable(output_file, data, sheetname="Data")
                end
                
            elseif format == "parquet"
                if data isa DataFrame
                    Parquet.write_parquet(output_file, data)
                end
                
            elseif format == "arrow"
                if data isa DataFrame
                    Arrow.write(output_file, data)
                end
                
            elseif format == "jld2"
                JLD2.save(output_file, "data", data)
                
            elseif format == "markdown"
                open(output_file, "w") do f
                    println(f, "# Data Export Report")
                    println(f, "\nGenerated: $(now())")
                    println(f, "\n## Data Summary")
                    if data isa DataFrame
                        println(f, "- Rows: $(nrow(data))")
                        println(f, "- Columns: $(ncol(data))")
                        println(f, "\n## Column Names")
                        for col in names(data)
                            println(f, "- $col")
                        end
                    end
                end
            end
            
            results[format] = "✅ Exported to $output_file"
            println("✅ Exported to $format format")
            
        catch e
            results[format] = "❌ Export failed: $(e)"
            println("⚠️  $format export failed: ", e)
        end
    end
    
    return results
end

# Generate comprehensive report
function generate_comprehensive_report(all_results::Dict{String, Any})
    report = Dict{String, Any}(
        "timestamp" => now(),
        "execution_summary" => Dict(
            "total_data_sources" => length(get(all_results, "data_sources", Dict())),
            "ai_models_used" => length(get(all_results, "ai_models", Dict())),
            "ml_models_trained" => length(get(all_results, "ml_results", Dict())),
            "files_processed" => length(get(all_results, "file_processing", Dict())),
            "parallel_processes" => get(all_results, "parallel_count", 0)
        ),
        "data_processing" => get(all_results, "data_processing", Dict()),
        "ai_analysis" => get(all_results, "ai_analysis", Dict()),
        "ml_results" => get(all_results, "ml_results", Dict()),
        "file_processing" => get(all_results, "file_processing", Dict()),
        "performance_metrics" => get(all_results, "performance", Dict()),
        "recommendations" => [
            "Consider scaling to distributed computing for larger datasets",
            "Implement caching for frequently accessed data",
            "Add more comprehensive error handling and logging",
            "Consider implementing real-time streaming analytics",
            "Add automated model retraining pipelines"
        ]
    )
    
    return report
end

# Export results in multiple formats
export_formats = ["json", "csv", "markdown", "excel"]

# Export processed data
for (source, df) in extracted_data
    if df isa DataFrame && nrow(df) > 0
        println("\n📤 Exporting $source data...")
        export_results = export_data(df, "processed_$(source)", export_formats)
        
        for (format, result) in export_results
            println("  $format: $result")
        end
    end
end

# Generate and export comprehensive report
println("\n📋 Generating comprehensive report...")
all_results = Dict(
    "data_sources" => extracted_data,
    "ai_models" => ai_models,
    "ml_results" => ml_results,
    "file_processing" => file_processing_results,
    "parallel_count" => length(data_sources),
    "performance" => Dict("parallel_processing" => parallel_results)
)

comprehensive_report = generate_comprehensive_report(all_results)

# Export report in multiple formats
report_export_results = export_data(comprehensive_report, "comprehensive_report", export_formats)

println("\n📊 Report Generation Summary:")
for (format, result) in report_export_results
    println("  $format: $result")
end

println("\n🎉 Julia ETL Processing Complete!")
println("📁 Check the current directory for exported files and reports.")
```
