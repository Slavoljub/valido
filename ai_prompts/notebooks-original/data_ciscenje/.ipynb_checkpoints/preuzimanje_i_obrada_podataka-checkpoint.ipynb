```python
import requests
from bs4 import BeautifulSoup

url = "https://www.mfin.gov.rs/propisi"
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'html.parser')
```

```python
def izdvoji_sekcije(soup):
    linkovi = []
    for link in soup.find_all('a', href=True):
        tekst = link.get_text(strip=True)
        if 'акти' in tekst.lower() or 'закони' in tekst.lower():
            linkovi.append((tekst, link['href']))
    return linkovi

sekcije = izdvoji_sekcije(soup)
for naziv, link in sekcije:
    print(f"Naslov: {naziv}, Link: {link}")
```

```python
from urllib.parse import urljoin

for naziv, link in sekcije:
    puna_putanja = urljoin(url, link)  # Pravi potpuni URL
    print(f"Prikupljam podatke sa: {puna_putanja}")
    response = requests.get(puna_putanja)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Izdvoji podatke sa stranice
    tekstovi = [el.get_text(strip=True) for el in soup.find_all('p')]
    print(f"Podaci sa sekcije {naziv}: {tekstovi}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL glavne stranice
url = "https://www.mfin.gov.rs/propisi"

# Preuzimanje sadržaja stranice
response = requests.get(url)
html = response.content

# Kreiranje BeautifulSoup objekta za lakšu obradu HTML-a
soup = BeautifulSoup(html, 'html.parser')

# Provera
print("Sadržaj stranice preuzet.")
```

```python
# Funkcija za izdvajanje linkova
def izdvoji_sekcije(soup):
    linkovi = []
    for link in soup.find_all('a', href=True):  # Pronalaženje svih <a> tagova
        tekst = link.get_text(strip=True)  # Tekst linka
        if 'акти' in tekst.lower() or 'закони' in tekst.lower():  # Relevantni linkovi
            linkovi.append((tekst, link['href']))
    return linkovi

# Poziv funkcije
sekcije = izdvoji_sekcije(soup)

# Prikaz izvučenih linkova
for naziv, link in sekcije:
    print(f"Naslov: {naziv}, Link: {link}")
```

```python
from urllib.parse import urljoin  # Za pravljenje potpunog URL-a

# Prolazak kroz svaku sekciju
for naziv, link in sekcije:
    puna_putanja = urljoin(url, link)  # Kreiranje kompletnog URL-a
    print(f"Prikupljam podatke sa: {puna_putanja}")
    
    # Preuzimanje sadržaja stranice
    response = requests.get(puna_putanja)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Izdvajanje svih paragrafa sa stranice
    tekstovi = [el.get_text(strip=True) for el in soup.find_all('p')]
    
    # Prikaz prikupljenih podataka
    print(f"Podaci sa sekcije {naziv}:")
    for tekst in tekstovi:
        print(tekst)
```

```python
import json

# Skladištenje podataka u JSON format
svi_podaci = {}
for naziv, link in sekcije:
    puna_putanja = urljoin(url, link)
    response = requests.get(puna_putanja)
    soup = BeautifulSoup(response.content, 'html.parser')
    tekstovi = [el.get_text(strip=True) for el in soup.find_all('p')]
    svi_podaci[naziv] = tekstovi

# Snimanje podataka u JSON fajl
with open("propisi_podaci.json", "w", encoding="utf-8") as f:
    json.dump(svi_podaci, f, ensure_ascii=False, indent=4)

print("Podaci uspešno sačuvani u propisi_podaci.json")
```

```python
# Funkcija za čišćenje podataka
def ocisti_podatke(podaci):
    ocisceni_podaci = []
    for red in podaci:
        red = red.strip()  # Uklanjanje vodećih i pratećih razmaka
        if red and len(red) > 5:  # Preskakanje praznih i kratkih redova
            ocisceni_podaci.append(red)
    return ocisceni_podaci

# Čitanje preuzetih podataka iz JSON fajla
import json

with open("propisi_podaci.json", "r", encoding="utf-8") as f:
    svi_podaci = json.load(f)

# Čišćenje podataka
svi_podaci_ocisceni = {}
for sekcija, tekstovi in svi_podaci.items():
    svi_podaci_ocisceni[sekcija] = ocisti_podatke(tekstovi)

# Provera čišćenih podataka
for sekcija, tekstovi in svi_podaci_ocisceni.items():
    print(f"\nSekcija: {sekcija}")
    for tekst in tekstovi[:3]:  # Prikaz samo prva tri reda za svaku sekciju
        print(f" - {tekst}")
```

```python
# Generisanje podataka za fine-tuning
fine_tuning_podaci = []

for sekcija, tekstovi in svi_podaci_ocisceni.items():
    for tekst in tekstovi:
        fine_tuning_podaci.append({
            "prompt": f"Šta kaže sekcija {sekcija} o ovom pitanju?",
            "completion": f" {tekst}"
        })

# Snimanje podataka u JSONL fajl
with open("fine_tuning_podaci.jsonl", "w", encoding="utf-8") as f:
    for red in fine_tuning_podaci:
        f.write(json.dumps(red, ensure_ascii=False) + "\n")

print("Podaci za fine-tuning su sačuvani u fine_tuning_podaci.jsonl")
```
