```python
from tabulate import tabulate
import pandas as pd

# Učitavanje podataka
# Ideja: Učitavamo prethodno očišćen skup podataka iz CSV fajla koji predstavlja ulazne fakture (KUF).
kuf = pd.read_csv('../../data/looping/kuf_csv/kuf_cleaned.csv')


# Selektovanje samo važnih kolona za pregled
# Ovde se biraju samo kolone koje su bitne za analizu: broj dokumenta, datum, kupac i iznos.
kolone = ['Broj dokumenta', 'Datum dokumenta', 'Naziv kupca', 'Ukupan iznos sa PDV-om', 'Pretplaćeni PDV odbitak']
kuf_prikaz = kuf[kolone].head(10)

# Prikaz tabele
# Cilj: Vizuelno prikazati podatke na jasan način pomoću stilizovane tabele.
from IPython.display import display

display(kuf.style.set_table_styles(
    [{'selector': 'th', 'props': [('font-size', '12pt'), ('text-align', 'center')]},
     {'selector': 'td', 'props': [('font-size', '10pt'), ('text-align', 'center')]}]
).set_properties(**{'border': '1px solid black'}))
```

```python
# Pretvaranje celog KUF-a u tekstualni format
# Funkcija uzima jedan red (fakturu) i kreira opis u čitljivom formatu.
def formatiraj_fakturu(row):
    return f"Faktura broj {row['Broj dokumenta']} od {row['Naziv kupca']} ima ukupnu vrednost sa PDV-om {row['Ukupan iznos sa PDV-om']} RSD. Datum dokumenta je {row['Datum dokumenta']}, a pretplaćeni PDV iznosi {row['Pretplaćeni PDV odbitak']} RSD."

# Kreiramo tekstualne opise za sve redove
# Primena funkcije na svaki red u tabeli pomoću liste.
tekstualni_opisi = [formatiraj_fakturu(row) for _, row in kuf.iterrows()]

# Prikaz celog teksta
# Svaki opis se prikazuje u čitljivom formatu.
for opis in tekstualni_opisi:
    print(opis)
```

```python
import openai

# Postavljanje OpenAI API ključa
# Ovo je neophodno za autentifikaciju prilikom pozivanja OpenAI API-ja.
import openai

# Postavi API ključ
openai.api_key = "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"
# Primer jedne fakture iz KUF-a
# Simulacija ulaznih podataka za analizu anomalija
primer_fakture = "Faktura broj 307-2024 od KOPILOVIC & KOPILOVIC AD ima ukupnu vrednost sa PDV-om 1300.0 RSD. Datum dokumenta je 7/22/2024, a ..."

# Inicijalizacija odgovora
response_text = ""  # Varijabla u kojoj se skladišti kompletan odgovor modela
continue_generation = True  # Kontrola petlje za generisanje odgovora

# Petlja za generisanje kompletnog odgovora
while continue_generation:
    # Slanje upita modelu
    # Model `gpt-3.5-turbo` analizira zadatu fakturu i odgovara
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # Naziv modela
        messages=[
            {"role": "system", "content": "Ti si analitičar faktura koji identifikuje anomalije."},  # Kontekst modela
            {"role": "user", "content": f"Analiziraj sledeću fakturu i reci da li ima anomalija: {primer_fakture}"}  # Pitanje korisnika
        ],
        max_tokens=200,  # Maksimalan broj tokena po odgovoru (ograničenje veličine odgovora)
        temperature=0.5  # Parametar za kontrolu kreativnosti odgovora
    )

    # Dodavanje generisanog dela odgovora u `response_text`
    # Ovo omogućava kontinuirano dodavanje teksta ukoliko odgovor nije završen u prvom pokušaju
    response_text += response['choices'][0]['message']['content']

    # Provera završetka odgovora
    # Provera da li odgovor završava sa tačkom, upitnikom ili uzvičnikom (kraj rečenice)
    if response_text.strip().endswith((".", "?", "!")):
        continue_generation = False  # Zaustavljanje petlje ako je odgovor kompletan

# Prikaz odgovora modela
# Nakon dobijanja celokupnog odgovora, prikazujemo ga korisniku
print("ValidoAI:")
print(response_text)
```

```python
import csv
import json

# Putanje do fajlova
input_csv_path = "../../data/looping/kuf_csv/kuf_cleaned.csv"
output_jsonl_path = "kuf_cleaned_openai_focused.jsonl"

# Relevantne kolone za analizu
relevant_columns = ["Broj dokumenta", "Ukupan iznos sa PDV-om", "Osnovica nabavke", "Iznos PDV-a"]

try:
    # Otvaranje CSV fajla za čitanje
    with open(input_csv_path, mode='r', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        data = []

        # Prolaz kroz redove CSV fajla
        for row in csv_reader:
            # Kreiranje JSONL strukture za svaki red
            user_content = (
                f"Analiziraj sledeću fakturu i reci da li ima anomalija: "
                f"Faktura broj {row['Broj dokumenta']} sa ukupnom vrednošću {row['Ukupan iznos sa PDV-om']} RSD, "
                f"osnovicom {row['Osnovica nabavke']} i PDV-om {row['Iznos PDV-a']} RSD."
            )
            
            jsonl_entry = {
                "messages": [
                    {"role": "system", "content": "Ti si analitičar faktura koji identifikuje anomalije na osnovu osnovice, PDV-a i ukupne vrednosti."},
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": "Odgovor se očekuje na osnovu pravila validacije podataka."}
                ]
            }
            data.append(jsonl_entry)

    # Pisanje u JSONL format
    with open(output_jsonl_path, mode='w', encoding='utf-8') as jsonl_file:
        for entry in data:
            json.dump(entry, jsonl_file)
            jsonl_file.write("\n")

    print(f"JSONL fajl je uspešno kreiran: {output_jsonl_path}")

except Exception as e:
    print(f"Došlo je do greške: {e}")
```

```python
import pandas as pd
import json

# Učitaj CSV fajl
csv_file_path = "kuf_cleaned_with_errors.csv"  # Putanja do vašeg CSV fajla
output_file_path = "kuf_cleaned_with_errors.jsonl"  # Putanja za izlazni JSONL fajl

# Učitaj CSV u pandas DataFrame
data = pd.read_csv(csv_file_path)

# Kreiraj JSONL fajl
with open(output_file_path, 'w') as jsonl_file:
    for _, row in data.iterrows():
        # Kreiraj JSON zapis za svaku fakturu
        jsonl_entry = {
            "messages": [
                {"role": "system", "content": "Ti si analitičar faktura koji identifikuje anomalije."},
                {"role": "user", "content": f"Analiziraj sledeću fakturu i reci da li ima anomalija: "
                                             f"Faktura broj {row['Broj dokumenta']} sa ukupnom vrednošću {row['Ukupan iznos sa PDV-om']} RSD, "
                                             f"osnovicom {row['Osnovica nabavke']} i PDV-om {row['Iznos PDV-a']} RSD."},
                {"role": "assistant", "content": "Odgovor se očekuje na osnovu pravila validacije podataka."}
            ]
        }
        # Zapiši u JSONL fajl
        jsonl_file.write(json.dumps(jsonl_entry) + '\n')

print(f"JSONL fajl je kreiran: {output_file_path}")
```

```python
from IPython.display import display, HTML
import openai
import ipywidgets as widgets

# Postavljanje API ključa za OpenAI
import openai

# Postavi API ključ
openai.api_key = "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"
# Funkcija za prikaz poruka u stilizovanom HTML formatu
def display_message(role, content):
    # Korisnikova poruka (poravnata desno, plava pozadina)
    if role == "user":
        display(HTML(f"""
        <div style="display: flex; justify-content: flex-end; margin: 10px;">
            <div style="padding: 10px; background-color: #e6f7ff; border-radius: 8px; width: auto; max-width: 60%; color: black;">
                <strong>Klijent:</strong> {content}
            </div>
        </div>
        """))
    # Odgovor asistenta (poravnata levo, siva pozadina)
    elif role == "assistant":
        display(HTML(f"""
        <div style="display: flex; justify-content: flex-start; margin: 10px;">
            <div style="padding: 10px; background-color: #f9f9f9; border-radius: 8px; width: auto; max-width:60%%; color: black;">
                <strong>ValidoAI:</strong> {content}
            </div>
        </div>
        """))
       

# Funkcija za slanje poruke OpenAI API-ju i prikaz odgovora
def send_message(model, user_message):
    # Definiše uloge i pravila za sistem i korisnika
    messages = [
        {"role": "system", "content": "Ti si analitičar faktura koji identifikuje anomalije. Pravila validacije uključuju: "
                                       "1. Ukupna vrednost = Osnovica + PDV. "
                                       "2. Osnovica mora biti veća od 0 osim u specifičnim slučajevima (npr. izuzeti računi). "
                                       "3. PDV mora biti pozitivan broj ukoliko postoji osnovica. "
                                       "4. Datumi dokumenta i knjiženja moraju biti logički konzistentni."},
        {"role": "user", "content": user_message}
    ]

    # Prikazuje korisničku poruku u UI
    display_message("user", user_message)

    # OpenAI API poziv za generisanje odgovora
    response = openai.ChatCompletion.create(
        model=model,  # Ime modela koji se koristi
        messages=messages  # Poruke koje se šalju API-ju
    )

    # Izvlači odgovor asistenta iz odgovora API-ja
    assistant_message = response['choices'][0]['message']['content']
    # Prikazuje odgovor asistenta u UI
    display_message("assistant", assistant_message)

# Funkcija koja kreira UI za interaktivni chat pomoću ipywidgets
def chat_with_model(model):
    # Tekstualni unos za pitanje korisnika
    input_box = widgets.Text(
        placeholder="Unesite pitanje vezano za fakturu...",  # Tekstualni nagoveštaj
        layout=widgets.Layout(width="80%",border_radius="5px")
    )
    send_button = widgets.Button(
    description="Pošalji",
    button_style="primary",
    tooltip="Kliknite da pošaljete pitanje",
    icon="paper-plane",  # Samo strelica
    layout=widgets.Layout(width="15%")  # Manje dugme
)


    # Oblast za prikaz poruka
    output_area = widgets.Output()

    # Funkcija koja se poziva kada se klikne na dugme
    def on_send_clicked(b):
        user_message = input_box.value.strip()  # Uzimanje unetog teksta
        if user_message:  # Provera da li je polje za unos prazno
            with output_area:  # Prikaz u izlaznoj oblasti
                send_message(model, user_message)
            input_box.value = ""  # Brisanje sadržaja iz polja za unos

    # Povezivanje dugmeta sa funkcijom za slanje
    send_button.on_click(on_send_clicked)

    # Stilizacija inputa i dugmeta
    input_and_button = widgets.HBox([input_box, send_button], layout=widgets.Layout(
        display="flex",
        justify_content="space-between",
        background_color="rgb(37, 150, 190)",  # Boja za pretragu
        padding="10px",
        border_radius="8px"
    ))

    
    # Prikaz celokupnog UI-ja (input + izlaz)
    display(widgets.VBox([input_and_button, output_area]))

# Pokretanje chata sa specifičnim modelom
chat_with_model("ft:gpt-3.5-turbo-1106:robot:kuf-cleaned-looping-with-errors:Al4OCyfI")
```
