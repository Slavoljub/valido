```python
import matplotlib.pyplot as plt
import seaborn as sns

# Stilizacija grafika
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 10))

# Podgraf 1: Status faktura
plt.subplot(2, 2, 1)
status_counts = cleaned_data['Status e-fakture'].value_counts()
sns.barplot(x=status_counts.index, y=status_counts.values, palette="crest")
plt.title("Distribucija statusa faktura")
plt.ylabel("Broj faktura")
plt.xlabel("Status")

# Podgraf 2: Iznos faktura (raspodela)
plt.subplot(2, 2, 2)
sns.histplot(cleaned_data['Iznos'], bins=15, color="orange", kde=True)
plt.title("Raspodela iznosa faktura")
plt.xlabel("Iznos (RSD)")
plt.ylabel("Broj faktura")

# Podgraf 3: Aktivnost po mesecima
plt.subplot(2, 1, 2)
mesec_counts = cleaned_data['Mesec'].value_counts().sort_index()
sns.barplot(x=mesec_counts.index, y=mesec_counts.values, palette="viridis")
plt.title("Broj faktura po mesecima")
plt.xlabel("Mesec")
plt.ylabel("Broj faktura")

# Finalni prikaz
plt.tight_layout()
plt.show()
```
