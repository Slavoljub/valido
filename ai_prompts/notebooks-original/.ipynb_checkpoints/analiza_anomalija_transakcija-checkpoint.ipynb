```python
import pandas as pd

# Učitavanje CSV fajla
file_path = '/home/malac/Desktop/MIT/projekat/validoAI/data/kif_looping_automation.csv'
data = pd.read_csv(file_path)

# Preimenovanje kolona
data.rename(columns={
    'br': 'Redni broj',
    'perobr': 'Period obračuna',
    'brdok': 'Broj dokumenta',
    'obj': 'Objekat',
    'datdok': 'Datum dokumenta',
    'datknj': 'Datum knjiženja',
    'kupac': 'ID kupca',
    'nazivkupca': 'Naziv kupca',
    'ulica': 'Adresa kupca',
    'broj': 'Broj adrese',
    'mesto': 'Mesto',
    'pib': 'PIB',
    'matbroj': 'Matični broj',
    'uknaksapdv': 'Ukupan iznos sa PDV-om',
    'oslprom_spo': 'Osnovica oslobođena SPO',
    'oslprom_bpo': 'Osnovica oslobođena BPO',
    'izvosl': 'Izvoz oslobođen PDV-a',
    'izvnosl': 'Iznos oslobođen PDV-a',
    'osnovicaopstastopa': 'Osnovica opšte stope PDV-a',
    'porezopstastopa': 'Porez opšte stope PDV-a',
    'osnovicaposebnast': 'Osnovica posebne stope',
    'porezposebnastopa': 'Porez posebne stope',
    'ukuprobezpdv': 'Ukupno bez PDV-a',
    'promspobezpdv': 'Promet SPO bez PDV-a'
}, inplace=True)

# Čuvanje rezultata u novi fajl
output_path = '/home/malac/Desktop/MIT/projekat/validoAI/data/kif_cleaned.csv'
data.to_csv(output_path, index=False)

print(f"Kolone su uspešno preimenovane i fajl je sačuvan na: {output_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje podataka
file_path = '/home/malac/Desktop/MIT/projekat/validoAI/data/kuf_cleaned.csv'
data = pd.read_csv(file_path)
# Provera osnovnih informacija
print("Osnovne informacije o podacima:")
print(data.info())

# Identifikacija anomalija
# Transakcije sa negativnim vrednostima
negativne_transakcije = data[(data['Ukupan iznos sa PDV-om'] < 0) | (data['Vrednost bez PDV-a'] < 0)]
print(f"Broj negativnih transakcija: {len(negativne_transakcije)}")

# Transakcije sa neobično velikim iznosima (preko 1,000,000 RSD)
velike_transakcije = data[data['Ukupan iznos sa PDV-om'] > 1_000_000]
print(f"Broj transakcija sa velikim iznosima: {len(velike_transakcije)}")

# Vizualizacija anomalija
plt.scatter(data['Redni broj'], data['Ukupan iznos sa PDV-om'], alpha=0.5, label='Sve transakcije')
plt.scatter(velike_transakcije['Redni broj'], velike_transakcije['Ukupan iznos sa PDV-om'], color='red', label='Velike transakcije')
plt.xlabel('Redni broj')
plt.ylabel('Ukupan iznos sa PDV-om')
plt.legend()
plt.title('Anomalije u transakcijama')
plt.show()

# Čuvanje anomalija u poseban CSV fajl
anomalije = pd.concat([negativne_transakcije, velike_transakcije])
anomalije.to_csv('putanja/do/vas/novi_fajl/anomalije.csv', index=False)

print("Anomalije su sačuvane u fajl: anomalije.csv")
```
