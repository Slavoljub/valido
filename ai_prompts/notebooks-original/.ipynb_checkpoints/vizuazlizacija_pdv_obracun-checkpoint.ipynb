```python
import pandas as pd

# Učitavanje očisćenih podataka
kpr_cleaned_data = pd.read_csv('ocisceni_kpr_podaci.csv')  # KPR: Evidencija ulaznih računa za PDV odbitak
kir_cleaned_data = pd.read_csv('ocisceni_kir podaci.csv')  # KIR: Evidencija izlaznih računa za PDV naplatu

# Provera podataka - prikazujemo prvih nekoliko redova za vizualni pregled
print("KPR Podaci: Evidencija ulaznih računa")
display(kpr_cleaned_data.head())  # Za proveru ulaznih računa

print("KIR Podaci: Evidencija izlaznih računa")
display(kir_cleaned_data.head())  # Za proveru izlaznih računa

# Dodajemo lepši tabelarni prikaz podataka
from IPython.display import display

# Prikaz KPR
print("Prikaz KPR Podataka")
display(kpr_cleaned_data.style.set_caption("Tabela: KPR - Ulazni Računi").highlight_max(axis=0))

# Prikaz KIR
print("Prikaz KIR Podataka")
display(kir_cleaned_data.style.set_caption("Tabela: KIR - Izlazni Računi").highlight_max(axis=0))

```

```python
from tabulate import tabulate

# Prikaz KPR podataka
print("KPR Podaci: Evidencija ulaznih računa")
print(tabulate(kpr_cleaned_data, headers='keys', tablefmt='pretty'))

# Prikaz KIR podataka
print("KIR Podaci: Evidencija izlaznih računa")
print(tabulate(kir_cleaned_data, headers='keys', tablefmt='pretty'))
```

```python
# Prečišćavanje podataka - uklanjanje suvišnih tačaka i konvertovanje u float
# Ovde uklanjamo tačke koje nisu deo decimalnog formata i zamenjujemo zapetu sa tačkom
kpr_cleaned_data['PDV_Odbitak'] = (
    kpr_cleaned_data['PDV_Odbitak']
    .str.replace('.', '', regex=True)  # Uklanja sve tačke
    .str.replace(',', '.', regex=True)  # Zamenjuje zapetu tačkom
    .astype(float)  # Konvertuje u numerički format (float)
)

kir_cleaned_data['PDV_Naplacen'] = (
    kir_cleaned_data['PDV_Naplacen']
    .str.replace('.', '', regex=True)  # Uklanja sve tačke
    .str.replace(',', '.', regex=True)  # Zamenjuje zapetu tačkom
    .astype(float)  # Konvertuje u numerički format (float)
)

# Obračun ukupnog odbitnog PDV-a iz KPR (ulazni računi)
total_pdv_odbitak = kpr_cleaned_data['PDV_Odbitak'].sum()

# Obračun ukupnog naplaćenog PDV-a iz KIR (izlazni računi)
total_pdv_naplacen = kir_cleaned_data['PDV_Naplacen'].sum()

# Izračunavanje salda PDV-a (Naplaćeni PDV - Odbitni PDV)
pdv_saldo = total_pdv_naplacen - total_pdv_odbitak

# Prikaz rezultata za proveru
{
    "Ukupan odbitni PDV (KPR)": total_pdv_odbitak,  # Ukupno odbitni PDV iz ulaznih računa
    "Ukupan naplaćeni PDV (KIR)": total_pdv_naplacen,  # Ukupno naplaćeni PDV iz izlaznih računa
    "Saldo PDV-a": pdv_saldo  # Konačna razlika PDV-a (za povraćaj ili uplatu državi)
}
```

```python
import matplotlib.pyplot as plt

# Priprema podataka za vizualizaciju
labels = ['Ukupan odbitni PDV (KPR)', 'Ukupan naplaćeni PDV (KIR)', 'Saldo PDV-a']
values = [total_pdv_odbitak, total_pdv_naplacen, pdv_saldo]

# Kreiranje bar grafika
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['blue', 'green', 'red'])
plt.title('Vizualizacija PDV Obračuna', fontsize=16)
plt.ylabel('Vrednosti (RSD)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Dodavanje vrednosti na vrh svakog bara
for i, v in enumerate(values):
    plt.text(i, v + (0.05 * max(values)), f'{v:.2f}', ha='center', fontsize=12, color='black')

# Prikaz grafika
plt.tight_layout()
plt.show()
```
