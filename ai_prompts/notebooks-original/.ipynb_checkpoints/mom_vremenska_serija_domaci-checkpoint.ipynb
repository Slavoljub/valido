```python
import pandas as pd
from IPython.display import display

# Učitavanje podataka
file_path = '../data/mom21/ocisceni_csv/cleaned_domaci.csv'
data = pd.read_csv(file_path)

# Filtriranje relevantnih kolona
relevant_columns = ['Datum Knjizenja', 'Mesto', 'Osnovica Opsta Stopa', 
                    'Porez Opsta Stopa', 'Ukupno Bez PDV-a', 'Prihodi']
filtered_data = data[relevant_columns]

# Funkcija za stilizaciju tabele
def style_table(df):
    return (df.style
            .set_table_styles(
                [
                    {'selector': 'thead th',
                     'props': [('background-color', '#1f4e79'),
                               ('color', 'white'),
                               ('font-weight', 'bold'),
                               ('text-align', 'center')]},
                    {'selector': 'tbody td',
                     'props': [('border', '1px solid black'),
                               ('color', '#333333')]}
                ]
            )
            .apply(lambda x: ['background-color: #e6f0fa' if i % 2 == 0 else 'background-color: white'
                              for i in range(len(x))], axis=0)
            .format({
                'Osnovica Opsta Stopa': 'RSD {:,.2f}',
                'Porez Opsta Stopa': 'RSD {:,.2f}',
                'Ukupno Bez PDV-a': 'RSD {:,.2f}',
                'Prihodi': 'RSD {:,.2f}'
            })
            .set_properties(subset=['Osnovica Opsta Stopa', 'Porez Opsta Stopa', 
                                    'Ukupno Bez PDV-a', 'Prihodi'],
                            **{'text-align': 'right'})
            .set_caption("Stilizovani podaci domaćeg prometa")
            .hide(axis="index"))

# Stilizovanje i prikaz podataka
styled_table = style_table(filtered_data)
display(styled_table)
```

```python
import pandas as pd
from IPython.display import display

# Učitavanje podataka
file_path = '../data/mom21/ocisceni_csv/cleaned_domaci.csv'
data = pd.read_csv(file_path)

# Zamenjujemo NaN vrednosti sa "--"
data.fillna('--', inplace=True)

# Filtriranje relevantnih kolona
relevant_columns = ['Datum Knjizenja', 'Mesto', 'Osnovica Opsta Stopa', 
                    'Porez Opsta Stopa', 'Ukupno Bez PDV-a', 'Prihodi']
filtered_data = data[relevant_columns]

# Funkcija za stilizaciju tabele
def style_table(df):
    return (df.style
            .set_table_styles(
                [
                    {'selector': 'thead th',
                     'props': [('background-color', '#1f4e79'),
                               ('color', 'white'),
                               ('font-weight', 'bold'),
                               ('text-align', 'center'),
                               ('width', '240px')]},  # Širina kolona
                    {'selector': 'tbody td',
                     'props': [('border', '1px solid black'),
                               ('color', '#333333'),
                               ('width', '200px')]}  # Širina za sadržaj
                ]
            )
            .apply(lambda x: ['background-color: #e6f0fa' if i % 2 == 0 else 'background-color: white'
                              for i in range(len(x))], axis=0)
            .format({
                'Osnovica Opsta Stopa': 'RSD {:,.2f}',
                'Porez Opsta Stopa': 'RSD {:,.2f}',
                'Ukupno Bez PDV-a': 'RSD {:,.2f}',
                'Prihodi': 'RSD {:,.2f}'
            })
            .set_properties(subset=['Osnovica Opsta Stopa', 'Porez Opsta Stopa', 
                                    'Ukupno Bez PDV-a', 'Prihodi'],
                            **{'text-align': 'right'})
            .set_caption("Stilizovani podaci domaćeg prometa")
            .hide(axis="index"))

# Stilizovanje i prikaz podataka
styled_table = style_table(filtered_data)
display(styled_table)
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose

# Pretpostavljamo da 'vremenska_serija' sadrži vremensku seriju sa vremenskim indeksom
# Primer: vremenska_serija = pd.Series(data, index=pd.date_range(start='2020-01-01', periods=len(data), freq='D'))

# Izračunavanje pokretnog proseka
window_size = 5  # Veličina prozora za pokretni prosek
vremenska_serija_sredina = vremenska_serija.rolling(window=window_size).mean()

# Dekompozicija vremenske serije
result = seasonal_decompose(vremenska_serija, model='additive', period=12)  # 'period' zavisi od učestalosti podataka

# Kreiramo DataFrame za Plotly (transformacija u long format)
df_plot = pd.DataFrame({
    'Datum': vremenska_serija.index,
    'Original': vremenska_serija,
    'Pokretni Prosek': vremenska_serija_sredina,
    'Trend': result.trend,
    'Sezonalnost': result.seasonal,
    'Rezidual': result.resid
})

# Transformišemo DataFrame u long format
df_long = pd.melt(df_plot, id_vars=['Datum'], var_name='Komponenta', value_name='Vrednost')

# Koristimo Plotly Express za vizualizaciju (faceted area plot)
fig = px.area(df_long, 
              x='Datum', 
              y='Vrednost', 
              color='Komponenta', 
              facet_col='Komponenta',  # Facet po komponentama vremenskih serija
              facet_col_wrap=2,  # Dve komponente u svakom redu
              title='Vremenska serija - Prihodi i Komponente',
              labels={"Datum": "Datum Knjizenja", "Vrednost": "Prihodi (RSD)"})

# Dodatna stilizacija
fig.update_layout(
    title='Vremenska serija - Prihodi i Komponente',
    xaxis_title='Datum Knjizenja',
    yaxis_title='Prihodi (RSD)',
    legend_title='Legenda',
    template='plotly_dark',  # Tamna pozadina sa svetlim linijama
    font=dict(color='white'),  # Belo slovo za bolji kontrast
    showlegend=True,  # Prikazujemo legendu
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Transparentna pozadina za grafikon
    paper_bgcolor='rgba(0, 0, 0, 0)',  # Transparentna pozadina za papir
    yaxis=dict(tickangle=45),  # Rotiramo oznake na y-osi za bolju preglednost
    height=800,  # Povećavamo visinu panela za bolji raspored
    title_x=0.5,  # Centriranje naslova grafikona
    title_y=0.95,  # Pozicija naslova na vertikalnoj osi
    legend=dict(
        x=1,           # Pomeramo legendu na desnu stranu
        xanchor='left',  # Poravnanje legende po levoj ivici
        y=0,           # Postavljamo legendu na dno
        yanchor='bottom' # Poravnanje legende po donjoj ivici
    )
)

fig.show()
```
