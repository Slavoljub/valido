```python
import pandas as pd
from IPython.display import display
import os
import sys
import openai
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.express as px

# Učitavanje podataka
file_path = '../../data/mom21/promet/cleaned_domaci.csv'
data = pd.read_csv(file_path)

# Zamenjujemo NaN vrednosti sa "--"
data.fillna('--', inplace=True)

# Filtriranje relevantnih kolona
relevant_columns = ['Datum Knjizenja', 'Mesto', 'Osnovica Opsta Stopa', 
                    'Porez Opsta Stopa', 'Ukupno Bez PDV-a', 'Prihodi']
filtered_data = data[relevant_columns]

# Funkcija za stilizaciju tabele
def style_table(df):
    return (df.style
            .set_table_styles(
                [
                    {'selector': 'thead th',
                     'props': [('background-color', '#1f4e79'),
                               ('color', 'white'),
                               ('font-weight', 'bold'),
                               ('text-align', 'center'),
                               ('width', '240px')]},  # Širina kolona
                    {'selector': 'tbody td',
                     'props': [('border', '1px solid black'),
                               ('color', '#333333'),
                               ('width', '200px')]}  # Širina za sadržaj
                ]
            )
            .apply(lambda x: ['background-color: #e6f0fa' if i % 2 == 0 else 'background-color: white'
                              for i in range(len(x))], axis=0)
            .format({
                'Osnovica Opsta Stopa': 'RSD {:,.2f}',
                'Porez Opsta Stopa': 'RSD {:,.2f}',
                'Ukupno Bez PDV-a': 'RSD {:,.2f}',
                'Prihodi': 'RSD {:,.2f}'
            })
            .set_properties(subset=['Osnovica Opsta Stopa', 'Porez Opsta Stopa', 
                                    'Ukupno Bez PDV-a', 'Prihodi'],
                            **{'text-align': 'right'})
            .set_caption("Stilizovani podaci domaćeg prometa")
            .hide(axis="index"))

# Stilizovanje i prikaz podataka
styled_table = style_table(filtered_data)
display(styled_table)
```

```python
from IPython.display import display, HTML
import openai
import ipywidgets as widgets

# Postavljanje OpenAI API ključa
openai.api_key = "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"

# ID modela
models = {
    "Analiza podataka": "ft:gpt-3.5-turbo-1106:robot:mom-vremenska-serija-jace-epohe:AloWlFVs",
    "Objašnjenje funkcionalnosti": "ft:gpt-3.5-turbo-1106:robot:validoai-finetuning:AgiIvcSo"
}

# Funkcija za prikaz poruka u stilizovanom chatu
def display_message(role, content):
    """Prikazuje poruku sa određenim stilom za korisnika ili ValidoAI."""
    if role == "user":
        display(HTML(f"""
        <div style="display: flex; justify-content: flex-end; margin: 10px;">
            <div style="padding: 10px; background-color: #e6f7ff; border-radius: 8px; width: auto; max-width: 60%; color: black;">
                <strong>Klijent:</strong> {content}
            </div>
        </div>
        """))
    elif role == "assistant":
        display(HTML(f"""
        <div style="display: flex; justify-content: flex-start; margin: 10px;">
            <div style="padding: 10px; background-color: #f9f9f9; border-radius: 8px; width: auto; max-width: 60%; color: black;">
                <strong>ValidoAI:</strong> {content}
            </div>
        </div>
        """))

# Funkcija za OpenAI API poziv i prikaz odgovora
def send_message(model, user_message):
    """Komunicira sa OpenAI modelom i prikazuje odgovor."""
    messages = [
        {"role": "system", "content": f"Ti si ValidoAI, fokusiran na {model}."},
        {"role": "user", "content": user_message}
    ]
    
    # Prikaz korisničke poruke
    display_message("user", user_message)
    
    # API poziv OpenAI modelu
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages
    )
    
    # Prikaz odgovora asistenta
    assistant_message = response['choices'][0]['message']['content']
    display_message("assistant", assistant_message)

# Interaktivni chat sa ipywidgets
def interactive_chat():
    """Kreira interaktivni chat sa unosom i dugmetom za slanje."""
    dropdown_menu = widgets.Dropdown(
        options=['Analiza podataka', 'Objašnjenje funkcionalnosti'],
        value='Analiza podataka',  # Podrazumevana vrednost
        description="ValidoAI:",
        style={'description_width': '100px'},  # Širina opisa
        layout=widgets.Layout(
            width="25%", 
            height="30px",  # Manja visina dropdown-a
        )
    )
    
    input_box = widgets.Text(
        placeholder="Postavite pitanje vezano za podatke...",
        layout=widgets.Layout(
            width="60%",
            height="30px",  # Manja visina input polja
            margin="0px 10px 0px 10px"  # Pomeranje input polja na desno
        )
    )
    send_button = widgets.Button(
        description="Pošalji",
        tooltip="Kliknite da pošaljete pitanje",
        icon="paper-plane",  # Strelica za slanje
        style={'button_color': '#258EBE'},  # Plava boja dugmeta
        layout=widgets.Layout(
            width="10%", 
            height="30px",  # Manja visina
            margin="0px 0px 0px 5px"  # Razmak između dugmeta i inputa
        )
    )

    output_area = widgets.Output()

    def on_send_clicked(b):
        user_message = input_box.value.strip()
        selected_model = dropdown_menu.value
        if user_message:
            with output_area:
                send_message(models[selected_model], user_message)
            input_box.value = ""  # Resetovanje input box-a

    send_button.on_click(on_send_clicked)

    # Kombinacija dropdown-a, input-a i dugmeta
    dropdown_and_button = widgets.HBox(
        [dropdown_menu, input_box, send_button],
        layout=widgets.Layout(
            display="flex",
            justify_content="space-between",
            align_items="center",  # Poravnanje na sredinu
            padding="10px",
        )
    )
    
    # Prikaz chat okruženja
    display(widgets.VBox([dropdown_and_button, output_area]))

# Pokretanje chata
interactive_chat()
```

```python
import pandas as pd
from IPython.display import display

# Pretvaranje CSV u DataFrame
file_path = '../../data/mom21/promet/cleaned_domaci.csv'
try:
    data = pd.read_csv(file_path)
except FileNotFoundError:
    display("CSV fajl nije pronađen. Proverite putanju do fajla.")

# Zamenjujemo NaN vrednosti sa "--"
data.fillna('--', inplace=True)

# Filtriranje relevantnih kolona
relevant_columns = ['Datum Knjizenja', 'Mesto', 'Osnovica Opsta Stopa',
                    'Porez Opsta Stopa', 'Ukupno Bez PDV-a', 'Prihodi']
filtered_data = data[relevant_columns]

# Provera numeričkih kolona i njihova konverzija
for col in ['Osnovica Opsta Stopa', 'Porez Opsta Stopa', 'Ukupno Bez PDV-a', 'Prihodi']:
    filtered_data[col] = pd.to_numeric(filtered_data[col], errors='coerce')

# Funkcija za stilizaciju tabele
def style_table(df):
    return (df.style
            .set_table_styles(
                [
                    {'selector': 'thead th',
                     'props': [('background-color', '#1f4e79'),
                               ('color', 'white'),
                               ('font-weight', 'bold'),
                               ('text-align', 'center'),
                               ('width', '240px')]},  # Širina kolona
                    {'selector': 'tbody td',
                     'props': [('border', '1px solid black'),
                               ('color', '#333333'),
                               ('width', '300px')]}  # Širina za sadržaj
                ]
            )
            .apply(lambda x: ['background-color: #e6f0fa' if i % 2 == 0 else 'background-color: white'
                              for i in range(len(x))], axis=0)
            .format({
                'Osnovica Opsta Stopa': 'RSD {:,.2f}',
                'Porez Opsta Stopa': 'RSD {:,.2f}',
                'Ukupno Bez PDV-a': 'RSD {:,.2f}',
                'Prihodi': 'RSD {:,.2f}'
            })
            .set_properties(subset=['Osnovica Opsta Stopa', 'Porez Opsta Stopa',
                                    'Ukupno Bez PDV-a', 'Prihodi'],
                            **{'text-align': 'right'})
            .set_caption("Stilizovani podaci domaćeg prometa")
            .hide(axis="index"))

# Stilizovanje i prikaz podataka
styled_table = style_table(filtered_data)
display(styled_table)
```
