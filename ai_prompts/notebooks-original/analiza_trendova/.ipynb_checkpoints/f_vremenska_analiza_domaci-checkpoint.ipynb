```python
import pandas as pd
import plotly.express as px
from statsmodels.tsa.seasonal import seasonal_decompose

def vremenska_analiza_domaci(file_path):
    """
    Funkcija za analizu vremenske serije domaćeg prometa.

    Parametri:
        file_path (str): Putanja do CSV fajla sa podacima.

    Vraća:
        fig (plotly.graph_objects.Figure): Grafikon vremenske serije sa komponentama.
    """
    # Učitavanje i priprema podataka
    cleaned_data = pd.read_csv(file_path)
    cleaned_data['Datum Knjizenja'] = pd.to_datetime(cleaned_data['Datum Knjizenja'])
    vremenska_serija = pd.Series(cleaned_data['Prihodi'].values, index=cleaned_data['Datum Knjizenja'])

    # Izračunavanje pokretnog proseka
    window_size = 5  # Veličina prozora za pokretni prosek
    vremenska_serija_sredina = vremenska_serija.rolling(window=window_size).mean()

    # Dekompozicija vremenske serije
    result = seasonal_decompose(vremenska_serija, model='additive', period=12)

    # Kreiranje DataFrame-a za vizualizaciju
    df_plot = pd.DataFrame({
        'Datum': vremenska_serija.index,
        'Original': vremenska_serija,
        'Pokretni Prosek': vremenska_serija_sredina,
        'Trend': result.trend,
        'Sezonalnost': result.seasonal,
        'Rezidual': result.resid
    })

    # Transformacija u long format za Plotly
    df_long = pd.melt(df_plot, id_vars=['Datum'], var_name='Komponenta', value_name='Vrednost')

    # Vizualizacija sa Plotly
    fig = px.area(df_long,
                  x='Datum',
                  y='Vrednost',
                  color='Komponenta',
                  facet_col='Komponenta',
                  facet_col_wrap=2,
                  title='Vremenska serija - Prihodi i Komponente',
                  labels={"Datum": "Datum Knjizenja", "Vrednost": "Prihodi (RSD)"})

    # Stilizacija grafikona
    fig.update_layout(
        title='Vremenska serija - Prihodi i Komponente',
        xaxis_title='Datum Knjizenja',
        yaxis_title='Prihodi (RSD)',
        legend_title='Legenda',
        template='plotly_dark',
        font=dict(color='white'),
        showlegend=True,
        plot_bgcolor='rgba(0, 0, 0, 0)',
        paper_bgcolor='rgba(0, 0, 0, 0)',
        height=800,
        title_x=0.5,
        title_y=0.95,
        legend=dict(
            x=1,
            xanchor='left',
            y=0,
            yanchor='bottom'
        )
    )

    return fig
```
