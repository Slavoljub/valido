```python
import sys
import os
import xml.etree.ElementTree as ET

# Apsolutna putanja do root direktorijuma tvog projekta
root_path = '/home/malac/Desktop/MIT/projekat/validoAI'

# Dodavanje root direktorijuma i src foldera u sys.path
sys.path.append(root_path)  # Dodaj root projekat
sys.path.append(os.path.join(root_path, 'src'))  # Dodaj src folder

# Uvoz funkcije iz openai_models.py
from src.models.openai_models import load_pppd_data

# Putanja do PPPD fajlova
data_folder_path = os.path.join(root_path, 'data/mom21/pppd')

# Provera foldera sa PPPD fajlovima
if not os.path.exists(data_folder_path):
    raise FileNotFoundError(f"Folder sa PPPD fajlovima ne postoji: {data_folder_path}")

xml_files = [f for f in os.listdir(data_folder_path) if f.endswith('.xml')]
if not xml_files:
    raise FileNotFoundError(f"Nema XML fajlova u folderu: {data_folder_path}")

print(f"Pronađeni XML fajlovi: {xml_files}")

# Funkcija za učitavanje podataka iz XML fajlova
def load_pppd_data(folder_path):
    """
    Učitava XML fajlove iz zadatog foldera i vraća listu podataka.
    """
    data = []

    for file in os.listdir(folder_path):
        if file.endswith(".xml"):
            file_path = os.path.join(folder_path, file)
            try:
                tree = ET.parse(file_path)
                root = tree.getroot()
                
                # Čitamo podatke o prijavi
                for prijava in root.findall(".//{http://pid.purs.gov.rs}PodaciOPrijavi"):
                    oznaka = prijava.find("{http://pid.purs.gov.rs}KlijentskaOznakaDeklaracije").text
                    period = prijava.find("{http://pid.purs.gov.rs}ObracunskiPeriod").text
                    datum = prijava.find("{http://pid.purs.gov.rs}DatumPlacanja").text
                    
                    # Dodajemo podatke u listu
                    data.append({
                        "Oznaka deklaracije": oznaka,
                        "Obračunski period": period,
                        "Datum plaćanja": datum
                    })
            except Exception as e:
                print(f"Greška prilikom obrade fajla {file}: {e}")
    
    return data

# Testiranje učitavanja podataka
def test_load_pppd_data():
    try:
        data = load_pppd_data(data_folder_path)
        if data:
            print("Podaci uspešno učitani:")
            for row in data[:5]:  # Prikazujemo prvih 5 redova
                print(row)
        else:
            print("Nema podataka u XML fajlovima.")
    except FileNotFoundError as e:
        print(f"Greška: {e}")
    except Exception as e:
        print(f"Neočekivana greška: {e}")

if __name__ == "__main__":
    # Testiranje učitavanja podataka iz XML fajlova
    test_load_pppd_data()
```

```python
import json
import os
import sys

# Apsolutna putanja do root direktorijuma tvog projekta
root_path = '/home/malac/Desktop/MIT/projekat/validoAI'

# Dodavanje root direktorijuma i src foldera u sys.path
sys.path.append(root_path)  # Dodaj root projekat
sys.path.append(os.path.join(root_path, 'src'))  # Dodaj src folder

# Uvoz funkcije
from src.models.openai_models import process_chat_with_employee_info, load_pppd_data

# Putanja do JSONL fajla i foldera sa PPPD podacima
file_path = os.path.join(root_path, 'data/mom21/prompts/pppd_prompts.jsonl')
data_folder_path = os.path.join(root_path, 'data/mom21/pppd')

def extract_employee_questions(file_path):
    """
    Izvlači pitanja koja potencijalno sadrže ime i prezime zaposlenog.
    """
    employee_questions = []
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line_number, line in enumerate(file, start=1):
                try:
                    data = json.loads(line)
                    if "messages" in data:
                        for message in data["messages"]:
                            if message["role"] == "user":
                                print(f"Poruka {line_number}: {message['content']}")
                                if any(word.istitle() for word in message["content"].split()):
                                    employee_questions.append(message["content"])
                except json.JSONDecodeError as e:
                    print(f"Greška u parsiranju linije {line_number}: {e}")
                    continue
    except FileNotFoundError:
        print(f"Fajl nije pronađen na putanji: {file_path}")
    except Exception as e:
        print(f"Neočekivana greška: {e}")
    
    print(f"Ekstrahovana pitanja: {employee_questions}")
    return employee_questions



def test_process_chat_with_employee_info():
    """
    Testira funkciju process_chat_with_employee_info sa pitanjima iz JSONL fajla.
    """
    # Učitaj podatke iz PPPD foldera
    data = load_pppd_data(data_folder_path)
    
    # Izvuci pitanja koja sadrže ime i prezime
    questions = extract_employee_questions(file_path)
    model_id = "ft:gpt-3.5-turbo-1106:robot:pppd:AkIWN4Sr"

    # Iteriraj kroz pitanja i testiraj funkciju
    for question in questions:
        print(f"Pitanje: {question}")
        response = process_chat_with_employee_info(question, data, model_id)
        print(f"Odgovor: {response}")
        print("-" * 50)

if __name__ == "__main__":
    test_process_chat_with_employee_info()
```
